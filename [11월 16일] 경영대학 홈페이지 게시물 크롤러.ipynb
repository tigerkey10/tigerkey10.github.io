```python
from selenium import webdriver
import selenium
import time
import os
import random
from bs4 import BeautifulSoup 
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
print('-'*80)
print('부산대학교 경영대학 - 학부 공지사항 제목 크롤러')
print('*'*80)
f_dir = '/Users/kibeomkim/Desktop/경영대학 공지사항 크롤링/'

now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' %(now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(f_dir+s+'-'+'경영대학 공지사항 크롤링 결과')
os.chdir(f_dir+s+'-'+'경영대학 공지사항 크롤링 결과')

cnt = input('몇 건을 수집하시겠습니까?') # 몇 개 게시물을 수집할 건가

fc_name = s+'-'+'경영대학 공지사항.csv'
fx_name = s+'-'+'경영대학 공지사항.xlsx'
```

```python
# 드라이버 정의
path = '/Users/kibeomkim/Desktop/chromedriver_intel/m1_chromedriver'
options= webdriver.ChromeOptions()
options.add_argument('User-Agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.0.2 Safari/605.1.15')
driver = webdriver.Chrome(path, options=options)
s_time = time.time()
driver.get('https://biz.pusan.ac.kr/biz/index..do')
time.sleep(3)
```

```python
# 학부 공지사항 더보기
driver.find_element_by_xpath('/html/body/section/div/article/div[1]/div/div[5]/div[2]/ul/li[1]/a').click()
time.sleep(3)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
total_no = soup.find('div', 'fnLeft').find('strong').get_text().replace('\n', "")
```

```python
number_list = []
title_list = []
writer_list = []
date_list = []
attached_list = []
clicks_list = []

no = 1
page_no = 1

while True : 
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    tr_list = soup.find('table', 'artclTable artclHorNum1').find('tbody').find_all('tr')

    for tr in tr_list : 
        # 게시글 번호 또는 종류
        number = tr.find('td', '_artclTdNum').get_text().replace('\n', '').strip()
        print('*-'*80)
        print(f'게시글 번호 또는 종류 : {number}')
        number_list.append(number)

        # 게시글 제목 
        title = tr.find('td', '_artclTdTitle').get_text().replace('\n', '').replace('\t', '')
        print(f'게시글 제목 : {title}')
        title_list.append(title)

        # 게시글 작성자 
        writer = tr.find('td', '_artclTdWriter').get_text().replace('\n', '').strip()
        print(f'게시글 작성자 : {writer}')
        writer_list.append(writer)

        # 게시글 작성일 
        date = tr.find('td', '_artclTdRdate').get_text().replace('\n', '').strip()
        print(f'게시글 작성일 : {date}')
        date_list.append(date)

        # 첨부파일 
        attached_file = tr.find('td', '_artclTdAtchFile').get_text().replace('\n', '').strip()
        print(f'첨부파일 갯수 : {attached_file}')
        attached_list.append(attached_file)

        # 조회수
        clicks = tr.find('td', '_artclTdAccess').get_text().replace('\n', '').strip()
        print(f'조회수 : {clicks}')
        clicks_list.append(clicks)

        no += 1 # 다음 순서 번호
        if no > int(cnt) : 
            break
    if no > int(cnt) : 
        driver.quit()
        break

    page_no += 1 # 다음 페이지 번호

    for a in range(2) : # 스크롤 다운
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        time.sleep(1.5)

    try : # 다음 페이지 번호 클릭
        driver.find_element_by_link_text(str(page_no)).click()
        time.sleep(3)
    except : 
        try : # '다음' 클릭
            driver.find_element_by_link_text('다음').click()
            time.sleep(3)
        except : # 페이지 끝. 더이상 게시물 없음. 종료
            print('다음 페이지가 없습니다')
            driver.quit()
            break
```

```python
business_board = pd.DataFrame()
business_board['번호'] = number_list
business_board['게시글 제목'] = title_list
business_board['게시글 작성자'] = writer_list
business_board['게시글 작성일자'] = date_list
business_board['첨부파일 갯수'] = attached_list
business_board['조회수'] = clicks_list

business_board.to_csv(fc_name, encoding='utf-8-sig')
business_board.to_excel(fx_name)

e_time = time.time()
total_time = e_time - s_time

print('='*80)
print(f'총 {total_no}건 중 {cnt} 건을 요청받아 {no-1} 건 만큼 수집했습니다')
print(f'총 소요시간은 {round(total_time,2)}초 입니다')
print('='*80)
```

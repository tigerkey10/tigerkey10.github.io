# 국가별 총액, 우수고객 분류하기 
- 1. groupby() 함수
- 형식 : df.groupby('분류할 칼럼 이름').함수()['분류 후 함수로 계산할 칼럼 선택']
    

```python
## 국가별 총액 구하기 
retail3.groupby('Country').sum()['Price']
```

```python
retail3.groupby('Country').sum()['Price'].sort_values(ascending=False)
```

# 우수고객 확인
- 구매횟수 기준
- 구매금액 기준

```python
# 고객 별 구매횟수 기준
retail3.groupby('CustomerID').count()['InvoiceNo'].sort_values(ascending=False)
```

```python
# 고객 별 구매금액 기준
retail3.groupby('CustomerID').sum()['Price'].sort_values(ascending=False)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm # 로지스틱 회귀분석 logit 함수 호출
from sklearn.model_selection import train_test_split # test값과 train값 나누기
from sklearn.preprocessing import StandardScaler # 전처리-표준화 함수

from sklearn.linear_model import LogisticRegression # 머신러닝 관점에서 로지스틱 회귀분석 - 알고리즘

from sklearn.metrics import confusion_matrix # 정오표 (모델 예측이 얼마나 정확했나? / 오분류 갯수는? )
from sklearn.metrics import accuracy_score # 모델 예측 정확도 
```

```python

y = df['Personal Loan']
x = df.drop('Personal Loan', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1, stratify=y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
stdsc = StandardScaler()
```

```python
x_train.iloc[:,:] = stdsc.fit_transform(x_train.iloc[:,:])
x_test.iloc[:,:] = stdsc.transform(x_test.iloc[:,:])
```

# 1. 데이터마이닝 관점에서 로지스틱 회귀분석 실행하기

```python

import statsmodels.api as sm
logit_model = sm.Logit(y_train, x_train).fit()
logit_model.summary()
```

```python
logit_model.params.sort_values(ascending=False) #로짓값들이다. 오즈비로 변환해서 각 독립변수가 확률에 어떻게 영향 미치는지 보자. 
```

```python
cat = logit_model.params
np.exp(cat).sort_values(ascending=False)
# income의 예 : 
# 2.017이 1보다 크기 때문에 양의 방향으로 확률에 영향 미친다.
# income이 1단위 증가할 때, 종속변수 y 가 1에 속할 확률이 2.017배 증가한다. (또는 101.7% 증가한다. )
```

# 2. 11주차 학습-  로지스틱 회귀분석 실습코드 작성 & 연습하기 

```python
# 머신러닝 관점에서 로지스틱 회귀분석 모델 세우기 & 예측 및 성능 평가하기 
logistic_ml = LogisticRegression(solver='liblinear', penalty='l2', C=0.001, random_state=1) # 머신러닝으로 학습시킬 모델 만들기 
logistic_ml.fit(x_train, y_train) # 모델에 TRAIN 데이터 넣어서 학습시키기.
```

```python
y_pred = logistic_ml.predict(x_test) # TRAIN 데이터 학습한 모델 가지고, TEST 데이터 넣어서 타겟 값 예측하기 
y_pred
y_pred_p = logistic_ml.predict_proba(x_test) # 타겟값이 0 또는 1로 나올 확률
y_pred_p
```

```python
y_test #실제값, 이제 이걸로 모델이 예측한 y_pred가 얼마나 정확했는지 볼 것이다. 
```

```python
# 정오표로 모델 예측이 얼마나 정확했는지 확인하기 
con = confusion_matrix(y_test, y_pred)
```

```python
confmat = pd.DataFrame(con)
confmat
print('오분류 갯수 : %d' %(y_test!=y_pred).sum()) #오분류 갯수 확인하기 
print('정확도 : %.3f' %accuracy_score(y_test, y_pred)) #accuracy_score 함수로 모델 예측 정확도 확인하기 
```

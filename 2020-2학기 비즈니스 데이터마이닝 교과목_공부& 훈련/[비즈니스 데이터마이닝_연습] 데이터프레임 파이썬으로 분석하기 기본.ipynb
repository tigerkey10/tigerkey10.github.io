# Chapter 6. 다중회귀분석
- OLS 함수로 모델링
- 전처리 -> x,y값 분리--> 상수항 추가--> 모델생성 및 학습 --> 해석

```python
import statsmodels.api as sm
```

```python
df.info()
df.describe()
```

```python
x = df[['Installs', 'Price', 'Rating']]
y = df['Reviews']
```

```python
x1 = sm.add_constant(x, has_constant='add')
x1
```

```python
multi_model = sm.OLS(y, x1).fit() #OLS함수는 회귀분석 실행해줌 / fit() 함수는 학습 명령임.
multi_model.summary() # 모델 생성하고 학습한 결과값을 보여라.
```

# 다시 책으로 돌아와서


```python
df = pd.read_csv('c:/bd092_data/ozone.csv')
df.head()
df.info() # 일조량, 오존에 공값 있다.
```

```python
df.describe()
```

```python
# 153개 뿐이기 때문에 레코드 잃으면 아깝다. 
# 공값 평균값으로 대체해주자.
df['Ozone'] = df['Ozone'].fillna(df['Ozone'].mean())
df.describe()
```

```python
df['Solar.R'] = df['Solar.R'].fillna(df['Solar.R'].mean())
df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
#1. x값과 y 값 분리한다.
x = df[['Solar.R', 'Wind', 'Temp']]
y = df['Ozone']
```

```python
#2. 상수항 추가한다.
import statsmodels.api as sm
x1 = sm.add_constant(x, has_constant='add')
```

```python
#3. 모델 생성한다.
multi_model = sm.OLS(y,x1)
#4. 학습시킨다.
result = multi_model.fit()
result.summary()
```

# chapter 7. 로지스틱 회귀분석
- 로지스틱 반응함수, 로짓, 그리고 오즈비를 활용
- x 값에 따라 집단 1에 속할 비율이 어떻게 바뀔까? 
- 학습데이터, 평가데이터 나누기
- 학습데이터, 평가데이터 표준화 하기
- 모델 생성에 Logit()함수 사용

- --> x,y값 분리 --> x,y를 다시 학습용, 평가용 데이터로 분리 --> 학습용 데이터, 평가용 데이터 표준화 시키기 + 학습용 데이터는 
- 표준화 하면서 학습 시키기 --> Logit 함수로 모델 생성 --> 로짓값으로 나온 회귀계수들 expo() 처리해서 오즈비로 바꾸기

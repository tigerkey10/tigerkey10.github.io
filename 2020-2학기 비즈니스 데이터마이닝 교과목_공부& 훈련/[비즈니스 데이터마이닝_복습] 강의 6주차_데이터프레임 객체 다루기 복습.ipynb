# 데이터 탐색(계산)

- 내장함수 mean(), std(), min(), max(), var() 사용

```python
cat['Quantity'].mean()#Quantity 칼럼에 들어있는 데이터값들의 평균
```

```python
cat['Quantity'].std() #Quantity 칼럼에 들어있는 데이터들의 표준편차
```

```python
cat['Quantity'].min() #해당 칼럼 데이터값들 중 최솟값
```

```python
cat['Quantity'].max() #해당 칼럼 데이터 값들 중 최댓값
```

```python
cat['Quantity'].var() #해당 칼럼 데이터값 분산값
```

결론 : 다양한 내장함수 활용으로 칼럼 데이터값들 분석이 가능하다

```python
cat.isnull()#cat 이라는 데이터프레임 내에 결측값 나타내주는 메소드
#결측값 = True / 결측 아닌 값 = False 
```

```python
cat.isnull().sum()#결측값 갯수만 한눈에 보기 쉽게 정리
```

# pd.notnull() 함수 사용해서 
null 값 제거하는 방법

```python
cat = cat[pd.notnull(cat['CustomerID'])] #칼럼 중심의 cat 데이터프레임에서 pd.notnull() 함수를 써서 <CultomerID 칼럼 null 값을 지워라>
```

```python
cat.isnull().sum()#cat 데이터프레임에서 null값 갯수만 나타내라
```

```python
##데이터프레임에서는 기본적으로 [] 연산자 쓴다. 
dog = pd.read_csv('c://bd092_data//NewOnlineRetail.csv')
dog.isnull().sum()
```

# 다시 복습하기 

```python
import pandas as pd
import numpy as np
```

```python
cat = pd.read_csv('c://bd092_data//NewOnlineRetail.csv')
## 전처리 위해 결측값 점검 & 수치 데이터 중 이상한 값 점검
#점겸 결과
#1. CustomerID 에 결측값 많음
#2. 수치 데이터 중 Quantity 값 최소가 마이너스 임
#<-- 제거 & 데이터 정제하자. 
```

```python
#1. customer ID 결측값 제거하자
cat.isnull().sum()
cat = cat[pd.notnull(cat['Description'])]
cat.isnull().sum()
cat = cat[pd.notnull(cat['CustomerID'])]
#데이터프레임은 column을 중심으로 움직인다.
cat.isnull().sum()
```

#2. 결측값은 다 없앴다. 두번째로 Quantity인데 마이너스인 값 제거하기 

```python
cat.describe()
```

```python
cat = cat[cat['Quantity']>0] #[] 연산자는 데이터프레임에서 '선택'을 의미한다
cat.describe()
```

```python
df = pd.read_csv('c:/bd092_data/BankPersonalLoan.csv')
df.drop(['ID', 'Experience', 'ZIP Code'], axis=1, inplace=True)
```

```python
# x,y 종속변수와 독립변수 분리하기 
x = df.drop('Personal Loan', axis=1)
y = df['Personal Loan']
```

```python
#학습용 / 평가용 70% : 30%으로 나누기
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=1, stratify=y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#표준화 작업
stdsc = StandardScaler() #함수의 활성화 <변수에 담아서 더 간단하게 쓰기>
#standard scaler 함수 사용하기 
x_train.iloc[:, :] = stdsc.fit_transform(x_train.iloc[:, :])
x_test.iloc[:, :] = stdsc.transform(x_test.iloc[:, :])
```

```python
# 이제 다시 모델 만들기 
log_model = sm.Logit(y_train, x_train)
result = log_model.fit()
```

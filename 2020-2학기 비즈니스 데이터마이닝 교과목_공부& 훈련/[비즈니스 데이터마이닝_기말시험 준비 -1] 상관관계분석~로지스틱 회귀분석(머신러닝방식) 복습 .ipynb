# 로지스틱 회귀분석

```python
import pandas as pd
import numpy as np 
import statsmodels.api as sm
```

```python
df = pd.read_csv('c://bd092_data/BankPersonalLoan.csv')
df.head()
df.info()
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df[df['Experience']>0]
df.describe()
df.info()
df.describe()
```

```python
df.head()
df.head()
df.info()
df.isnull().sum()
```

```python
df.head()
```

```python
df.describe()
```

```python
x = df.drop('Personal Loan', axis=1)
y = df['Personal Loan']
import statsmodels.api as sm
x1 = sm.add_constant(x, has_constant='add')
x1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1, stratify=y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
stdsc = StandardScaler()
x_train.iloc[:, :] = stdsc.fit_transform(x_train.iloc[:,:])
x_test.iloc[:,:] = stdsc.transform(x_test.iloc[:,:])
```

```python
logit_model = sm.Logit(y_train, x_train).fit()
logit_model.summary()
```

```python
cat = logit_model.params.sort_values(ascending=False)
np.exp(cat).sort_values(ascending=False)
```

# income 1단위 증가할 수록 집단 1 대출성공 그룹에 분류될 확률은 1.85배 증가한다. 
# 오즈비 값이 1보다 크기 때문에, 양의 방향으로 영향 미친다. 
# 1.85-1 = 0.85*100 = 85% 증가한다. 
# 따라서 income이 높으면 높을수록 고객이 대출성공 그룹에 분류될 확률이 높아진다. 
# income이 높은 고객을 선정해서 프로모션 해야 효율적이다. 
# 결과적으로 x값 레코드들 모델에 대입해서 p값 확률 구하고, 해당 레코드를 1 또는 0으로 분류한다. (0.5 기준)


```python
# 머신러닝 방식 로지스틱 회귀분석 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
logistic_model = LogisticRegression(solver='liblinear', penalty='l2', C=0.001, random_state=1)
logistic_model.fit(x_train, y_train)
```

```python
y_pred = logistic_model.predict(x_test)
y_pred
y_pred_p = logistic_model.predict_proba(x_test)
y_pred_p # 왼쪽은 0이 나올 확률, 오른쪽은 1이 나올 확률 
# 첫번째 예는 0.68이 0.5를 넘기때문에 y_pred를 0으로 분류함. 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confmat = pd.DataFrame(confusion_matrix(y_test, y_pred))
confmat
```

```python
#정오분류표 좀 더 보기 편하게 만들기 
#오분류 갯수
print('오분류갯수 : %d' %(y_test!=y_pred).sum())
#정확도 
print('정확도 : %.3f' %accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('c://bd092_data/BankPersonalLoan.csv')
df.head()
df.info()
df.describe()
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
x = df.drop('Personal Loan', axis=1)
y = df['Personal Loan']
import statsmodels.api as sm
x1 = sm.add_constant(x, has_constant='add')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1, stratify=y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, f1_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
tree_m = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=1)
tree_m.fit(x_train,y_train)
```

```python
y_pred = tree_m.predict(x_test)
y_pred
```

```python
y_pred_p = tree_m.predict_proba(x_test)
y_pred_p
tree_m.score(x_test, y_test)
```

```python
confmat = pd.DataFrame(confusion_matrix(y_test, y_pred))
confmat
```

```python
#오분류 갯수, 정확도, f1점수 
print('오분류 갯수 : %d' %(y_test!=y_pred).sum())
print('정확도 : %.3f' %accuracy_score(y_test, y_pred))
print('f1_score : %.3f' %f1_score(y_test, y_pred))
```

```python
#의사결정나무 시각화 하기 
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

feature_name = x.columns.tolist()
target_name = np.array(['No', 'Yes'])
```

```python
plt.figure(figsize=(18,15))

tree2 = plot_tree(tree_m,
                 feature_names = feature_name,
                 class_names = target_name,
                 filled=True,
                 rounded=True,
                 fontsize=14)
#income처럼 위쪽에 있는 x값일수록 중요한 값이다. 
#income이 <= 104.5인 값 기준으로 분할했다 .
#그렇게 분할하면 gini가 0.174다. (최소 gini 값)<--이게 사실 분할 기준이다.  
#왼쪽이 yes, 오른쪽이 no다. 
#분할되어 내려갈 수록 gini값이 낮아진다(불순도가 낮아진다.)
```

```python
from sklearn.pipeline import make_pipeline, Pipeline
asdf = make_pipeline(DecisionTreeClassifier())
asdf.get_params().keys()
```

```python
# 의사결정 나무 정지기준 찾는다 : depth어디까지 할건지 + 최종 leaf 최소 갯수는 몇개까지 할 건지
# 다시말해 '어디까지 분할할지' 찾는다. 
```

```python
#최적화 하기 

from sklearn.model_selection import GridSearchCV

params_range1 = range(1,11)
params_range2 = range(5, 26, 5)

params_grid = [{'decisiontreeclassifier__max_depth' : params_range1, 
               'decisiontreeclassifier__min_samples_leaf' : params_range2}]

clf_cv = GridSearchCV(estimator = asdf, 
                     param_grid = params_grid,
                     scoring='accuracy', 
                     n_jobs=-1,
                     cv=10)

clf_cv.fit(x_train, y_train)
```

```python
print(clf_cv.best_score_)
print(clf_cv.best_params_)
```

```python
tree_m_new = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=1, min_samples_leaf=10,  )
tree_m_new.fit(x_train, y_train)
```

```python
y_pred = tree_m_new.predict(x_test)
y_pred
tree_m_new.score(x_test, y_test)
```

```python
confmat = pd.DataFrame(confusion_matrix(y_test, y_pred))
confmat
```

```python
print('오분류 갯수 : %d' %(y_test!=y_pred).sum())
print('정확도 : %.3f' %accuracy_score(y_test, y_pred))
print('f1_score : %.3f' %f1_score(y_test, y_pred))
```

# 데이터프레임 칼럼과 row 선택하기 

데이터프레임은 column 선택이 기본값이다. 


데이터프레임에서 [] 연산자는 '선택'을 의미한다

```python
import pandas as pd
import numpy as np
cat = pd.read_csv('c://bd092_data//NewOnlineRetail.csv')
cat['InvoiceNo'] #[] 일때는 시리즈로 호출 (1개 열만 호출)
cat[['InvoiceNo']] #[[]] 일때는 데이터프레임으로 호출
cat[['InvoiceNo', 'Quantity']]
# 데이터프레임 기본 호출은 [[]]이다.
#슬라이싱 일때는 [ ] 원래대로 하나만 쓴다.
cat.loc[100:150, ['InvoiceNo']]
cat.loc[100, ['InvoiceNo']]

```

```python
cat.iloc[[100], [3,4,5]]
```

```python
cat.iloc[[100], [3,5,2]]
```

```python
cat.loc[100,['Quantity']]
cat.loc[[100],['Quantity']]
cat.iloc[[100], [2,5,6]]
cat.loc[[100,34,2342,25342,33,3], ['Quantity']]
cat.iloc[[33,55,234,242,121],[3,2,1,5]]
cat.iloc[[33],[4,2,1]]
cat.iloc[33,[4,2]]
#loc과 iloc 규칙 : 
#loc 과 iloc의 기본은 row 표현이다. <-- 뒤에 칼럼 값 더해주는 거는 
#사실 별 상관이 없다. 
#loc과 iloc 함수에서 row 번호 선택을 [33]로 하면 : 시리즈로 표현, + 해당 칼럼 추가 
#loc과 iloc 함수에서 row 번호 선택을 [[33]]으로 하면 : 데이터프레임으로 표현 + 해당 칼럼 추가 
#번호선택을 [[]]으로 할 경우, 여러 row 선택 가능 + 칼럼 추가 
#loc 일 때는 칼럼추가할 때 칼럼 이름만
#iloc 일 때는 칼럼추가할때 칼럼 순서만(0부터 시작 기준)
cat.iloc[[33],[3,2]]
```

```python
cat.loc[3:55]
#슬라이싱은 대괄호 갯수 상관 없이 알아서 데이터프레임으로 출력한다
cat.loc[3:77]
```

# 불리언을 이용해 특정 데이터 선택하기

```python
cat[(cat['UnitPrice']>=3) & (cat['Quantity']>=10)]
#cat 이라는 데이터 프레임에서 []를 선택해라. 
#뭘 선택하냐면 A,B 두 조건 모두를 만족시키는 값만 선택해라는 의미
```

```python
cat[(cat['Quantity']>=3) | (cat['UnitPrice']>=10)]
```

```python
cat['Price'] = cat['Quantity'] * cat['UnitPrice']
#데이터프레임은 모든게 칼럼 위주로 돌아간다
#cat.insert(3, 'Price2') <-- 오류난다. 이유 : 데이터프레임에 insert함수로 값 추가할 땐 해당 변수와 변수 값을 같이 넣어줘야 한다.
#cat.insert(2,'price2',cat['Quantity']*cat['UnitPrice'])
cat
```

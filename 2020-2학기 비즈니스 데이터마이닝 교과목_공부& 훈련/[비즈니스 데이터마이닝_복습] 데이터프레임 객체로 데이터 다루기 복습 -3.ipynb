# 데이터프레임 행과 열 선택하기 
- [] 대괄호 사용, 변수명[] 대괄호는 '변수에서 선택하라'는 뜻
- 데이터프레임에서는 []대괄호가 기본

```python
#행과 열 선택 예시
# retail[]
#retail 에서 [] 대괄호 안 내용 선택해라는 뜻
retail['InvoiceNo']
#대괄호 하나 = 열 하나 선택, 시리즈로 출력
retail[['InvoiceNo']]
#대괄호 둘 = 열 하나 또는 여러 개 선택, 데이터프레임으로 출력
retail[['InvoiceNo', 'UnitPrice']]
#retail['InvoiceNo', 'UnitPrice'] #열 하나 선택해서 시리즈 출력하란 뜻인데 열 두개 들어가서 에러
type(retail[['UnitPrice','InvoiceNo']])
```

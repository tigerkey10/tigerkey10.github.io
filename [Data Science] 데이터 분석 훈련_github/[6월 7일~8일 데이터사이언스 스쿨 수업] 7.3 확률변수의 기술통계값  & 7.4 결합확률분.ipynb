```python
# p.476
np.random.seed(15)
N = 7
data = np.sort(np.random.normal(size=N))[::-1]
data
```

```python
# 표본들의 표본평균 
mean = np.mean(data)
mean
```

```python
plt.figure(figsize = (10,2))
sns.rugplot(data, height=0.5, linewidth=4)

#x = np.linspace(-3,3,100)

plt.axvline(x=0, ls=':', c='b', linewidth=2, label='실제 기댓값')
plt.axvline(x=mean, ls='-.', c='r', linewidth=2, label='표본평균')
plt.legend()
plt.xlim(-2,2)
plt.title('데이터들이 모여 있는 곳 따라 왼쪽으로 편향 된 표본평균')
plt.show()
```

```python
distance_from_sample_mean = data - mean
distance_from_sample_mean # 편차 
```

# 표본분산 

```python
sample_variance = (distance_from_sample_mean**2).mean() # 표본분산 
sample_variance
```

# 표본분산은 대체적으로 확률변수 분산값보다 작게 나오는 경향 보인다. 
- 이 경우 확률변수 분산값은 1이었다. 

```python
# 편향오차 제거한 비편향 표본분산을 구해보자. 
(sample_variance*N)/(N-1)
```

# 비편향 표본분산 기댓값은 확률변수 분산값과 일치한다. 
- = 위 비편향 표본분산 값은 확률변수 분산값의 근삿값이 된다. 

```python
# 예제 
grades = ['A','B', 'C','D','E','F']
scores = pd.DataFrame([
    [1,2,1,0,0,0],
    [0,2,3,1,0,0],
    [0,4,7,4,1,0],
    [0,1,4,5,4,0],
    [0,0,1,3,2,0],
    [0,0,0,1,2,1]
], columns=grades, index=grades)

scores.columns.names=['X']
scores.index.names=['Y']

scores
```

# 위 데이터프레임 values들의 결합확률질량함수 값 구해보자. 

```python
pmf = scores/np.sum(scores.values)
pmf
```

```python
#heatmap으로 위 값들 시각화하기 
sns.heatmap(pmf, cmap=mpl.cm.bone_r , annot=True,
           xticklabels=['A','B','C','D','E','F'],
           yticklabels=['A','B','C','D','E','F'])
plt.title('결합확률질량함수 $pXY(x,y)$ 값들')
plt.tight_layout()
plt.show()
```

```python
import scipy as sp

# 두 연속확률변수의 기댓값 
mu = [70, 170]

# 두 연속확률변수의 공분산 
cov = [[150,140],
      [140,300]]

#다변수정규분포 - 결합확률밀도함수 값들 
rv = sp.stats.multivariate_normal(mu, cov)

xx = np.linspace(20,120,100)
yy = np.linspace(100,250,100)
XX,YY = np.meshgrid(xx,yy)
# 결합확률밀도함숫값 
ZZ = rv.pdf(np.dstack([XX,YY]))

plt.contour(XX,YY,ZZ)
plt.title('결합확률밀도함수')
plt.show()
```

```python
pmf1 = np.array([[1,2,4,2,1],
                [2,4,8,4,2],
                [4,8,16,8,4],
                [2,4,8,4,2],
                [1,2,4,2,1]])

# 두 이산확률변수의 결합확률분포 
pmf1 = pmf1/pmf1.sum()
pmf1
```

```python
# 주변확률 질량함수
pmf1_marginal_x = np.round(pmf1.sum(axis=0),2) #p(x)
pmf1_marginal_y = np.round(pmf1.sum(axis=1),2) #p(y)
pmf1x = pmf1_marginal_x * pmf1_marginal_y[:,np.newaxis] # 두 확률변수 X,Y 독립이라 pmf1과 똑같다.

plt.subplot(121)
sns.heatmap(pmf1, cmap=mpl.cm.bone_r, annot=True, square=True, linewidth=1, linecolor='k',
           cbar=False, xticklabels=pmf1_marginal_x, yticklabels=pmf1_marginal_y)
plt.title('서로 독립인 확률변수 X,Y의 결합확률 질량함수')

plt.subplot(122)
sns.heatmap(pmf1x, cmap=mpl.cm.bone_r, annot=True, square=True, linewidth=1, linecolor='k',
           cbar=False, xticklabels=pmf1_marginal_x, yticklabels=pmf1_marginal_y)
plt.title('두 확률변수 X,Y 주변확률질량함수 곱')
plt.show()
```

```python
pmf2 = np.array([[0,0,0,5,5],
                [0,5,5,5,5],
                [0,5,30,5,0],
                [5,5,5,5,0],
                [5,5,0,0,0]])
pmf2 = pmf2/np.sum(pmf2) # 결합확률 분포 

#주변확률 질량함수
pmf2_marginal_x = np.round(pmf2.sum(axis=0),2)
pmf2_marginal_y = np.round(pmf2.sum(axis=1),2)

plt.subplot(121)
sns.heatmap(pmf2, cmap=mpl.cm.bone_r, annot=True, square=True, linewidth=1,linecolor='k',
           cbar=False, xticklabels=pmf2_marginal_x, yticklabels=pmf2_marginal_y)
plt.title('상관관계인 X,Y 두 확률변수의 결합확률 질량함수')

plt.subplot(122)
pmf2x = pmf2_marginal_x*pmf2_marginal_y[:,np.newaxis]
sns.heatmap(pmf2x, cmap=mpl.cm.bone_r, annot=True, square=True, linewidth=1, 
           linecolor='k', 
           cbar=False,
           xticklabels=pmf2_marginal_x,
           yticklabels=pmf2_marginal_y)
plt.title('상관관계인 두 확률변수 주변확률 질량함수의 곱')
plt.tight_layout()
plt.show()
```

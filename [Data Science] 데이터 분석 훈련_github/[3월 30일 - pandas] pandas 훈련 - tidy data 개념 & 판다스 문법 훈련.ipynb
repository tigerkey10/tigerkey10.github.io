# 문제점 : 
- 깔끔한 데이터 공리 1번 위반 : 칼럼에 여러 변수 값들이 들어가 있다. (칼럼에는 유니크한 변수 하나만 들어갈 수 있다)
- 깔끔한 데이터 공리 1번 위반 : 인덱스에 여러 변수 값들이 들어가 있다. (변수는 칼럼에만 들어가야 한다)
- 표에 관측단위는 하나다. (깔끔한 데이터 공리 2번 만족)

```python
# 인덱스에 여러 변수 값들이 들어가 있는 문제 해결하기 - df.reset_index(drop=False)
df3_mi_mc = df3_mi_mc.reset_index(drop=False).rename(mapper={'level_0':'공원이름'}, axis=1).rename(mapper={'level_1':'구역이름'},axis=1)
```

```python
df3_mi_mc
```

```python
# 칼럼에 여러 변수 값들이 들어가 있는 문제 해결하기 - df.melt()
df3_mi_mc = df3_mi_mc.melt(id_vars=['공원이름','구역이름'],var_name=['종류','종류 세부구분'],value_name='공원에서 볼 수 있는 것들')
```

```python
df3_mi_mc.head(7) #깔끔한 데이터 
# - 칼럼에는 변수 한개, 각 변수는 유니크 해야 한다. 
# - 각 row는 유니크해야한다. 
# - 각 표에 관측단위는 하나만 존재해야 한다. 모두 만족.
```

```python
df3_mi_mc.drop_duplicates(inplace=True)
```

```python
park = df3_mi_mc #깔끔한 데이터 완성
park.head()
```

```python
# pd.concat - 2 #멀티 인덱스란 multi-hierachy 즉 층위가 여러 개인 인덱스를 말한다. 
df_add3 = pd.DataFrame([['코끼리','소나무','호수','다람쥐 동상','워터캠프']],
            columns=[['생물','생물','무생물','무생물','무생물'],
                    ['동물','식물','자연물','조형물','놀이기구']],
            index=pd.MultiIndex.from_tuples([['별님','구역5']]))
df_add3
```

```python
df3_mi_mc_concat = pd.concat([df3_mi_mc_concat,df_add3],axis=0,verify_integrity=True) #각각의 '값' 또한 유니크 해야한다.
```

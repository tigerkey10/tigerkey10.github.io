```python
t = np.linspace(-100, 500, 100) # -100 500   100  
F = t/360 #  
F[t<0] = 0 #    0       0  
F[t>360] = 1  #     360      1  
plt.plot(t, F) #     
plt.ylim(-0.1, 1.1) # y 
plt.xticks([0, 180, 360]) # x    
plt.title('') #   
plt.xlabel('$x$ ()') #x   
plt.ylabel('$F(x)$') #y   
plt.show()
```

```python
#6.4.4  
#  F(x)
def F(x) : 
    if x <= 0 : 
        P = 0
    if 0 < x <= 180 : 
        return (2/3)*(x/180)
    if 180 < x <= 360 : 
        return (2/3) + ((1/3)*((x-180)/180))
    if 360 < x : 
        return 1
F(270)
```

```python
P_list = []

for x in np.linspace(-100, 500, 1000) : 
    if x <= 0 : 
        P = 0 
        P_list.append(P)
    if 0 < x <= 180 : 
        P = (2/3)*(x/180)
        P_list.append(P)
    if 180 < x <= 360 : 
        P = (2/3) + ((1/3)*((x-180)/180))
        P_list.append(P)
    if 360 < x : 
        P = 1
        P_list.append(P)
```

```python
plt.plot(np.linspace(-100, 500, 1000), P_list) #     
plt.ylim(-0.1, 1.1) # y 
plt.xticks([0, 180, 360]) # x    
plt.title('  ') #   
plt.xlabel('$x$ ()') #x   
plt.ylabel('$F(x)$') #y   
plt.show()
```

```python
#6.4.5  
#  p(x)
t = np.linspace(-100, 500, 1000) # -100 500  1000  
p = np.gradient(P_list, 600/1000)
plt.plot(t, p)
plt.ylim(-0.0001, p.max()*1.1)
plt.xticks([0, 180, 360]) #x  0, 180, 360   
plt.title('  ')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.show()
```

```python
#6.4.5  
def p(x) : 
    if x <= 0 : 
        return 0
    if 0 < x <= 180 : 
        return (2/3)/180 #   # /    =  (=   )
    if 180 < x <= 360 : 
        return (1/3)/180 # /   =  (=     )
    if 360 < x : 
        return 0
p(181)
```

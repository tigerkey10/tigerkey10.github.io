# 기하학 관점에서의 벡터 
- 1. 공간 속 하나의 점
- 2. 원점에서 그 점을 잇는 화살표

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=18)
gray = {'facecolor' : 'gray'}
black = {'facecolor' : 'black'}
red = {'facecolor' : 'red'}
green = {'facecolor' : 'green'}
blue = {'facecolor' : 'blue'}
a = np.array([1,2])
plt.plot(0,0,'kp',ms=20)
plt.plot(a[0],a[1],'ro',ms=20)
plt.annotate('',xy=[-0.6,1.6], xytext=(0.2,0.7), arrowprops=gray)
plt.annotate('',xy=a,xytext=(0,0),arrowprops=black )
plt.annotate('',xy=a+[-1,1],xytext=(-1,1),arrowprops=black)
plt.annotate('',xy=a+[-1,1],xytext=(-1,1),arrowprops=black)
plt.text(0.35,1.15,'$a$')
plt.text(1.15,2.25,'$(1,2)$')
plt.text(-0.7, 2.1, '$a$')
plt.text(-0.9,0.6,'평행이동')
plt.xticks(np.arange(-2,4))
plt.yticks(np.arange(-1,4))
plt.xlim(-2.4, 3.4)
plt.ylim(-0.8,3.4)
plt.show()
```

```python
a = np.array([1,2])
np.linalg.norm(a) #벡터(화살표)의 길이(크기)
```

```python
# 단위벡터 : 놈 값이 1인 벡터 (길이가 1인 벡터)
a = np.array([1,0])
b = np.array([0,1])
c = np.array([1/np.sqrt(2),1/np.sqrt(2)])
np.linalg.norm(a),np.linalg.norm(b), np.linalg.norm(c)
```

```python
a = np.array([1,1])
b = np.array([-1,1])
b@a
a@b # 두 벡터는 서로 직교한다. + 두 벡터가 직교하면 두 벡터 내적도 0이다. 
```

```python
a = np.array([[4],
             [5],
             [2],
             [2]])
b = np.array([[4],
             [0],
             [2],
             [0]])
c = np.array([[2],
             [2],
             [0],
             [1]])
```

# 유클리드 거리 

```python
# 1. a-b 
_1 = a-b
print(_1)
np.linalg.norm(_1)
```

```python
# 2. a-c
_2 = a-c
print(_2)
np.linalg.norm(_2)
```

```python
# 3. b-c
_3 = b-c
print(_3)
np.linalg.norm(_3)
```

# b와 c가 가장 가깝다. a와 b가 가장 멀다. 

# 코사인 거리 

```python
1-((a.T@b)/(np.linalg.norm(a)*np.linalg.norm(b)))
```

```python
1-((a.T@c)/(np.linalg.norm(a)*np.linalg.norm(c)))
```

```python
1-((b.T@c)/(np.linalg.norm(b)*np.linalg.norm(c)))
```

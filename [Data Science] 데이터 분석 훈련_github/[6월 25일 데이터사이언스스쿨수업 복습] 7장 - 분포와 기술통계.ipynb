# 표본평균
- 다른 말 : 대푯값
- 역할 : 데이터 분포의 대략적 위치 나타낸다. (대략 어디쯤 위치한 분포인가?)
- 표본들 산술평균값

# 표본 중앙값
- 표본들 오름차순 정렬했을 때 가장 정 중앙에 위치한 값
- 표본 개수가 짝수 : (n+1)/2번째 값
- 표본 개수가 홀수 : n/2번째 값과 n/2+1번째 값의 산술평균


# 표본최빈값
- 가장 빈도수 높은 표본값
- 연속적 데이터 표본최빈값 구하는 법 

데이터들을 구간 - 구간으로 나눈다. 빈도수 가장 높게 나오는 구간의 대푯값(평균)을 전체 데이터의 표본최빈값으로 삼는다. 

```python
# 대푯값 계산 
print(f'표본평균 : {np.mean(X)}, 표본중앙값 {np.median(X)}')
```

```python
np.histogram(X, bins=bins) # 구간과 구간별 빈도수
```

```python
ns, _ = np.histogram(X, bins=bins)
m_bin = np.argmax(ns)
print(f'최빈구간 = {bins[m_bin]}~{bins[m_bin+1]}')
result = [i for i in X if i >= 0 and i <= 0.5]
print(f'표본 대푯값(평균) : {np.mean(result)}')
print(f'표본 중앙값 : {np.median(result)}')
```

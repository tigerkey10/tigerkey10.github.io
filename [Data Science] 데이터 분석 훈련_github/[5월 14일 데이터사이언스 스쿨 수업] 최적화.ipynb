# 최저점 근처에서 최적화 처음 시작할 수록, 
# 함수 호출 횟수가 줄어든다. 

```python
#연습문제 5.1.1
import scipy as sp
def f2(x) : 
    return (1-x[0])**2 + 100.0*(x[1]-x[0]**2)**2

def jac(x) : 
    return [2*x[0]-2+400*x[0]**3-400*x[1]*x[0], 200*x[1]-200*x[0]**2]


x0 = np.array([2, 0.3])

#1. 
result = sp.optimize.minimize(f2, x0, jac=jac)
result
```

# 전역 최적화 문제 

```python
x = sympy.symbols('x')
a = x**2-20*sympy.cos(x)
sympy.diff(a,)
```

```python
def f_global(x) : 
    return x**2-20*np.cos(x)
def f_global_p(x) : 
    return 2*x+20*np.sin(x)

x = np.linspace(-20, 20, 100)
plt.plot(x, f_global(x))
plt.scatter(0, f_global(0), s=200)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('전역 최적화 문제')
plt.show()
```

```python
for i in [7.5, 2.5, -2.5, -7.5] : 
    result1 = sp.optimize.minimize(f_global, i,)
    print(result1)
    print()
    result2 = sp.optimize.minimize(f_global, i, jac=f_global_p)
    print(result2)
    print()
    print('#'*100)
```

# 위 경우 jac=인자에 그레디언트 벡터 함수 넣어줬을 때(1차 도함수)

# 컨벡스 문제 
- 밑에서 봤을 때 볼록한 구간들만 잘라서 최적화
- 컨벡스 문제에서는 항상 전역 최저점 존재.

```python
def f_global(x) : 
    return x**2-20*np.cos(x)
def f_global_p(x) : 
    return 2*x+20*np.sin(x)

x = np.linspace(-7.5, -4, 100)
plt.plot(x, f_global(x))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('2차 도함수가 항상 0보다 큰 영역 1')
plt.show()
```

```python
def f_global(x) : 
    return x**2-20*np.cos(x)
def f_global_p(x) : 
    return 2*x+20*np.sin(x)

x = np.linspace(-2.5, 2.5, 100)
plt.plot(x, f_global(x))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('2차 도함수가 항상 0보다 큰 영역 2')
plt.show()
```

```python
def f_global(x) : 
    return x**2-20*np.cos(x)
def f_global_p(x) : 
    return 2*x+20*np.sin(x)

x = np.linspace(4, 7.5, 100)
plt.plot(x, f_global(x))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('2차 도함수가 항상 0보다 큰 영역 3')
plt.show()
```

```python
# 컨벡스 영역1에서 최적화 해보자. 
for i in x : 
    result = sp.optimize.minimize(f_global, i, jac=f_global_p)
print(result['fun'], '\n', result['x'])
```

```python
# 컨벡스 영역2에서 최적화 해보자. 
for i in x : 
    result = sp.optimize.minimize(f_global, i, jac=f_global_p)
print(result['fun'], '\n', result['x'])
```

```python
# 표본분포의 균형이 표본평균, 표본최빈값, 그리고 표본중앙값에 어떤 영향 미치는지 알아보자. 
np.random.seed(1)
x = np.random.normal(size=1000) # 시뮬레이션 횟수 : 1000번 # 1000개 표본 정규분포로 부터 추출
x = np.hstack([x,5*np.ones(50)]) #분포 균형 깨는 데이터 추가 
bins = np.linspace(-6,6,12*4+1)
ns, _ = np.histogram(x, bins=bins)

plt.figure(figsize=(10,5))
sns.distplot(x, bins=bins)
plt.title('분포의 균형은 기술통계값 변화에 어떤 영향 미치는가?')
plt.xlabel('$x$')

np.median(x)
np.mean(x)
sample_mode = 0.5*(bins[np.argmax(ns)]+bins[np.argmax(ns)+1])

plt.axvline(np.median(x), ls='--', c='k', label='표본중앙값')
plt.axvline(np.mean(x), ls='-',c='g', label='표본평균')
plt.axvline(sample_mode, ls='-.', c= 'r', label='표본최빈값')

plt.legend()
plt.show()
```

# 분포 좌우대칭이 깨지면 표본최빈값이 가장 영향 덜 받고, 표본 평균이 가장 영향 많이 받는다.

```python
np.random.seed(1)
x = np.random.normal(size=1000)

sns.distplot(x, bins=bins)
np.mean(x)
np.median(x)
ns, _ = np.histogram(x, bins=bins)

plt.axvline(np.mean(x), ls='--', c='k', label='표본평균')
plt.axvline(np.median(x), ls='-.',c='g', label='표본중앙값')
plt.axvline(0.5*(bins[np.argmax(ns)]+bins[np.argmax(ns)+1]), c='r',label='표본최빈값')
plt.legend()
plt.title('좌우대칭 깨지기 전 히스토그램')
plt.show()
```

```python
from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data.data, columns=[data.feature_names])
df['species'] = data.target
```

```python
# 7.1.2 연습문제 
#1.
X = df[df['species']==2]['petal length (cm)'].values
bins = np.arange(0,11)
sample_mean = np.mean(X)
sample_median = np.median(X)
ns,_ = np.histogram(X, bins=bins)
max_index = np.argmax(ns)
sample_mode = 0.5*(bins[max_index]+bins[max_index+1])

sns.distplot(X, bins=bins)
plt.axvline(sample_mean, ls='--',c='g', label='표본평균')
plt.axvline(sample_median, ls='-.',c='r',label='표본중앙값')
plt.axvline(sample_mode, ls='-', c='b',label='표본최빈값')
plt.legend()
plt.show()
```

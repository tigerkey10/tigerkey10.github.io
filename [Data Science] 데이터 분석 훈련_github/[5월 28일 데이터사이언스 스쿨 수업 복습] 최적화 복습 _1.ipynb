```python
# 2차원 함수 
def f2(x,y) : 
    return (1-x)**2+100*(y-x**2)**2

# 2차원 함수의 도함수 
def f2g(x,y) : 
    return np.array((-400*x*(-x**2+y)+2*x-2, -200*x**2+200*y))

xx = np.linspace(-4, 4, 800)
yy = np.linspace(-3,3, 600)
X, Y = np.meshgrid(xx, yy)
Z = f2(X, Y)

levels = np.logspace(-1, 3, 10)

plt.contour(X,Y,Z, alpha=0.2,  levels=levels) # alpha는 선명도, levels는 척도
plt.contour(X,Y,Z, colors='green', levels=levels, zorder=0)
#plt.plot(1,1, 'ro', markersize=10)

mu = 8e-4
s = 0.95 

x, y = -1,-1 # starting point

g1 = f2g(x,y)
for i in range(8500) : 
    g = f2g(x,y) # 그레디언트 벡터 값 
    plt.arrow(x, y, -2*mu*g[0], -2*mu*g[1], 
             head_width=0.04, head_length=0.04, fc='k', ec='k', lw=2)
    plt.plot(x,y, 'ro', markersize=3)
    x = x-mu*g[0]
    y = y-mu*g[1]
```

# 그레디언트 벡터 방향 : 기울기 가장 작아지는 방향.
# 근데 너무 많이 가면 - 가 된다. 내 목표는 기울기 0되는 지점
# $\mu$만큼만 이동. 그레디언트 벡터는 이 이동방향을 나타낸 것이다. 

# 진동현상 발생해서 최적화 비효율적으로 되는 경우 

```python
xx = np.linspace(0, 4, 800)
yy = np.linspace(0,3, 600)
X,Y = np.meshgrid(xx,yy)
Z = f2(X,Y)

levels = np.logspace(-1,4,20)

plt.contourf(X,Y,Z,alpha=0.2, levels=levels)
plt.contour(X,Y,Z,colors='green', levels=levels, zorder=0)
plt.plot(1,1,'ro', markersize=10)

mu = 1.8e-3
s = 0.95

x,y = 1.5,1.5
for i in range(10000) : 
    g = f2g(x,y)
    #plt.arrow(x,y,-s*mu*g[0], -s*mu*g[1],
             #head_width=0.04, head_length=0.04, fc='k', ec='k', lw=2)
    plt.plot(x,y, 'ro', markersize=5)
    x=x-mu*g[0]
    y = y-mu*g[1]
    
```

# 사이파이로 최적화 하기 

```python
import scipy as sp
# 목적함수 
def f1(x) : 
    return (x-2)**2+2

x0 = 0
result = sp.optimize.minimize(f1, x0)
print(result) #__str__() 메서드로 받았을 것이다. 
```

```python
def f2(x) : 
    return 2*(x-2)

result = sp.optimize.minimize(f1, x0, jac=f2)
print(result)
```

```python
# 2차원 다변수함수가 목적함수인 경우 
def f2(x) : 
    return (1-x[0])**2 +100*(x[1]-x[0]**2)**2

x0 = np.array([-2,-2])
result = sp.optimize.minimize(f2, x0)
print(result)
```

```python
# 전역 최적화 문제 <-- 컨벡스 문제로 풀어야 된다. 
def f_global(x) : 
    return x**2-20*np.cos(x)
x = np.arange(-10, 10, 0.1)
plt.plot(x, f_global(x))
plt.scatter(0, f_global(0), s=200)
plt.plot(0,f_global(0),'ro', markersize=10)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('전역 최적화 문제')
plt.show()
```

```python
# 초깃값이 적절하지 않아 전역 최저점 수렴 못 하는 경우 
result = sp.optimize.minimize(f_global, 7.5, )
print(result)
a = result.x
print(a)
```

# 사이파이로 등식 제한조건 있는 최적화 문제
# 계산하기 

```python
def f1array(x) : 
    return x[0]**2+x[1]**2

def eq_constraint(x) : 
    return x[0]+x[1]-1

sp.optimize.fmin_slsqp(f1array, np.array([1,1]), eqcons=[eq_constraint])
```

```python
def f2plt(x1,x2) : 
    return np.sqrt((x1-4)**2+(x2-2)**2)

x1 = np.linspace(-2, 5, 100)
x2 = np.linspace(-1.5, 3, 100)
X1, X2 = np.meshgrid(x1,x2)
Z = f2plt(X1,X2)

plt.contour(X1,X2, Z, colors='gray', 
           levels=np.arange(0.5, 5, 0.5)*np.sqrt(2))

# 제한조건의 상수
k = 0.1
ax = plt.gca()

x12 = np.linspace(-k, 0, 10)
x13 = np.linspace(0, k, 10)
ax.fill_between(x12, x12+k, -k-x12, color='g', alpha=0.5)
ax.fill_between(x13, x13-k, k-x13, color='g', alpha=0.5)


#최적점 위치 
def f2(x) : 
    return np.sqrt((x[0]-4)**2+(x[1]-2)**2)

def ieq_constraint(x) : 
    return np.atleast_1d(k-np.sum(np.abs(x)))
r = sp.optimize.fmin_slsqp(f2, np.array([0,0]), ieqcons=[ieq_constraint])
plt.plot(r[0],r[1], 'ro', markersize=5)
```

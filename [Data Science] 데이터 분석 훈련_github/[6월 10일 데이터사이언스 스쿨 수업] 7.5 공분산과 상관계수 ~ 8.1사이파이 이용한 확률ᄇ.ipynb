```python
# p.503 예제 
from sklearn.datasets import load_iris
import scipy as sp
X = load_iris().data

x1 = X[:,0] #꽃받침 길이
x2 = X[:,1] #꽃받침 폭
x3 = X[:,2] #꽃잎 길이
x4 = X[:,3] #꽃잎 폭
result = sp.stats.pearsonr(x1, x3)[0]
print('피어슨상관계수(표본상관계수) : %s'%result)
print("'표본분포가 오른쪽 위를 향한다'")
```

```python
#8.1.2 연습문제 
rv = sp.stats.norm(loc=1, scale=2) # 기댓값 1, 표준편차 2인 정규분포 객체 
sample_data = rv.rvs(size=1000, random_state=0)

#표본평균
print('표본평균 {}'.format(np.mean(sample_data)))

#표본분산 
print('편향 표본분산 {}'.format(np.var(sample_data)))
print()

#비편향 표본분산
print(f'비편향 표본분산 {np.var(sample_data, ddof=1)}')
```

# 변환 확률변수의 시뮬레이션 
## 변환 확률변수의 표본 분포가 어떻게 되는지, 직접 시험해보자.

```python
rv1 = sp.stats.uniform()
rv2 = sp.stats.uniform() #두 개 연속균등분포 객체 생성 

np.random.seed(0)
N=50000
x_1 = rv1.rvs(N) # 확률변수 1에서 무작위 표본 5만 개 추출
x_2 = rv2.rvs(N) # 확률변수 2에서 무작위 표본 5만 개 추출
x_3 = x_1 + x_2 # 변환 확률변수의 표본 5만 개 

plt.figure(figsize=(12,5))
plt.subplot(131)
sns.histplot(x_1, kde=True)
plt.title('균일분포')
plt.xlabel('표본값')
plt.xlim(-0.2, 2.2)

plt.subplot(132)
sns.histplot(x_2, kde=True)
plt.title('균일분포')
plt.xlabel('표본값')
plt.xlim(-0.2, 2.2)

plt.subplot(133)
sns.histplot(x_3, kde=True)
plt.title('변환 확률변수의 표본분포 - 균일분포 표본 합의 분포')
plt.xlabel('표본값')
plt.xlim(-0.2, 2.2)

plt.tight_layout()
plt.show()
```

```python
# 연습문제 8.1.3 
X = sp.stats.uniform()
N = 100000
result = [X.rvs(N) for i in range(10)]
converted_variable = np.array(result).sum(axis=0)

plt.figure(figsize=(12,5))
sns.histplot(converted_variable, kde=True)
plt.title('균등분포에서 시뮬레이션 1000번*10개 표본 생성해서 더한 결과')
plt.xlabel('표본값')
plt.show()
```

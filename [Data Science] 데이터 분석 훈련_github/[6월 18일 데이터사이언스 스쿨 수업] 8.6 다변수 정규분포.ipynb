# 다변수 정규분포는 
# N차원 벡터를 입력으로 받는 다변수함수다. 
### N차원 벡터 = 다변수 확률변수(벡터)

```python
mu = np.array([2,3])
cov = np.array([[1,0],[0,1]])

rv = sp.stats.multivariate_normal(mu, cov)
X = rv.rvs(20000) #확률변수벡터 표본
xx = np.linspace(-6,6,120)
yy = np.linspace(-1,6,150)
XX,YY = np.meshgrid(xx,yy)

plt.scatter(X[:,0],X[:,1],s=1) 
plt.contour(XX,YY,rv.pdf(np.dstack((XX,YY))))
plt.axis('equal')
plt.xlim(0,4)
plt.ylim(2,4)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('2차원 다변수 정규분포 예')
plt.show()
```

```python
# 예제 2 # 다변수 정규분포
mu = np.array([2,3])
cov = np.array([[2,3],[3,7]])

rv = sp.stats.multivariate_normal(mu, cov)
X = rv.rvs(20000) #표본 확률변수 벡터 추출

xx = np.linspace(-6,6,120)
yy = np.linspace(-6,6,150)
XX,YY = np.meshgrid(xx,yy) #그리드포인트 생성

plt.scatter(X[:,0],X[:,1], s=1)
plt.contour(XX,YY,rv.pdf(np.dstack([XX,YY])))
plt.axis('equal')
plt.xlim(0,4)
plt.ylim(2,4)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('2차원 다변수 정규분포 예시')
plt.show()
```

```python
mu = [2,3]
cov = [[4,3],[3,5]]
w, V = np.linalg.eig(cov)
print(f'고윳값 {w}')
print('고유벡터행렬')
print(f'{V}')
```

```python
plt.figure(figsize=(8,4))

d = dict(facecolor='k',edgecolor='k',width=2)

plt.subplot(121)
xx = np.linspace(-1,5,120)
yy = np.linspace(0,6,150)
XX,YY = np.meshgrid(xx,yy)
```

```python
rv1 = sp.stats.multivariate_normal(mu, cov)
plt.contour(XX,YY, rv1.pdf(np.dstack([XX,YY])))
plt.annotate('',xy=(mu+V[:,0]),xytext=mu,arrowprops=d)
plt.annotate('',xy=(mu+V[:,1]),xytext=mu, arrowprops=d)
#plt.scatter(2,3,s=10,c='k')
```

```python
plt.figure(figsize=(8,4))

d = dict(facecolor='k',edgecolor='k',width=2)

plt.subplot(121)
xx = np.linspace(-1,5,120)
yy = np.linspace(0,6,150)
XX,YY = np.meshgrid(xx,yy)
rv1 = sp.stats.multivariate_normal(mu, cov)
plt.contour(XX,YY,rv1.pdf(np.dstack([XX,YY])))
plt.annotate('',xy=(mu+V[:,0]),xytext=mu, arrowprops=d)
plt.annotate('',xy=(mu+V[:,1]),xytext=mu, arrowprops=d)
plt.axis('equal')

plt.subplot(122)
xx = np.linspace(-3,3,120)
yy = np.linspace(-3,3,150)
XX,YY = np.meshgrid(xx,yy)
rv2 = sp.stats.multivariate_normal([0,0],w)
plt.contour(XX,YY,rv2.pdf(np.dstack([XX,YY])))
plt.annotate('',xy=(np.array([1,0])),xytext=(0,0),arrowprops=d)
plt.annotate('',xy=(np.array([0,1])),xytext=(0,0),arrowprops=d)
plt.scatter(0,0,s=10,c='k')
```

```python
#8.6.1 연습문제 
mu = np.array([1,2])
cov = np.array([[4,-3],[-3,4]])

rv = sp.stats.multivariate_normal(mu, cov)

xx = np.linspace(-1,3,200)
yy = np.linspace(0,4,200)
XX,YY = np.meshgrid(xx,yy)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.contour(XX,YY,rv.pdf(np.dstack([XX,YY])))
plt.xlim(-5,6)
plt.ylim(-7.5,7.5)
plt.ylim(-4,7)
plt.xlim(-1,3)
plt.ylim(0,4)

plt.scatter(mu[0],mu[1],s=30,c='k')
plt.text(mu[0],mu[1],'$\mu$')
plt.title('2차원 다변수정규분포 확률밀도함수')

plt.annotate('',xy=(mu+2*V[:,0]),xytext=mu,arrowprops=d)
plt.annotate('',xy=(mu+1.5*V[:,1]),xytext=mu, arrowprops=d)

plt.subplot(122)
ld, V = np.linalg.eig(cov)
rv2 = sp.stats.multivariate_normal([0,0],ld)
xx = np.linspace(-10,10,200)
yy = np.linspace(-10,10,200)
XX,YY = np.meshgrid(xx,yy)
plt.contour(XX,YY,rv2.pdf(np.dstack([XX,YY])))
plt.xlim(-7.5, 7.5)
plt.ylim(-4,4)

plt.scatter(0,0,s=30,c='k')
plt.annotate('',xy=(0.5*ld[0]*np.array([1,0])),xytext=[0,0],arrowprops=d)
plt.annotate('',xy=(1.4*ld[1]*np.array([0,1])),xytext=[0,0],arrowprops=d)
plt.title('다변수정규분포 확률밀도함수 - 좌표변환 후')

plt.show()
```

```python
# 다시해보기 
mu = np.array([2,3])
cov = np.array([[1,0],[0,1]])

plt.subplot(121)
rv = sp.stats.multivariate_normal(mu, cov)
xx = np.linspace(-1.5,5,100)
yy = np.linspace(-2,7.5,200)
X,Y = np.meshgrid(xx,yy)

plt.contour(X,Y,rv.pdf(np.dstack([X,Y])))
ld, V = np.linalg.eig(cov)
plt.annotate('',xy=(mu+0.9*V[:,0]),xytext=mu, arrowprops=d)
plt.annotate('',xy=(mu+0.9*V[:,1]),xytext=mu, arrowprops=d)

plt.subplot(122)
rv2 = sp.stats.multivariate_normal([0,0],ld)
xx = np.linspace(-3,5,100)
yy = np.linspace(-5,5,200)
XX,YY = np.meshgrid(xx,yy)
plt.contour(XX,YY,rv2.pdf(np.dstack([XX,YY])))
plt.xlim(-3,3)
plt.annotate('',xy=(ld[0]*V[:,0]),xytext=[0,0], arrowprops=d)
plt.annotate('',xy=(ld[1]*V[:,1]),xytext=[0,0], arrowprops=d)
plt.title('좌표변환 후')
plt.show()
```

```python
# 3.4.1 연습문제 
B = np.array([[3,2,2],
             [2,3,-2]])
C = np.array([[2,4],
             [1,3],
             [0,0],
             [0,0]])

############
U, sg, VT = np.linalg.svd(B)
U1, sg1, VT1 = np.linalg.svd(C)

SG = np.diag(sg, -1)[1:]
SG1 = np.diag(sg1, 2)[:,2:]

print('특이분해 곱 결과 ')
print(U@SG@VT)
print()
print(U1@SG1@VT1)

###########
U, sg, VT = np.linalg.svd(B, full_matrices=False)
U1, sg1, VT1 = np.linalg.svd(C, full_matrices=False)

SG2 = np.diag(sg)
SG3 = np.diag(sg1)

print('-'*100)
print('특이분해 축소형 이용한 결과')
print(U@SG2@VT)
print()
print(U1@SG3@VT1)
```

# 테스트_ 
- 차원축소한 저차원벡터는 고차원벡터의 투영벡터를 좌표변환한 것이다!

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

X = load_iris()
X1 = X.data #붓꽃데이터

pca = PCA(n_components=1)
X1_low = pca.fit_transform(X1)
X2 = pca.inverse_transform(X1_low)
```

```python
X1_low[0]
```

```python
w = pca.components_ #주성분
w
```

```python
#투영벡터 구하자
a = X1[0]-pca.mean_
a@w.T@w #1차원 공간에 대한 투영벡터
VER = a@w.T@w
VER
```

# PCA 이해 정리

1. 근사 성능 높이기 위해서 데이터에서 공통분모 빼고 나머지를 벡터공간 투영한다. 

```python
((X1[0]-pca.mean_)@(w).T*(w)) 
```

2. 투영벡터는 아래처럼 표현할 수도 있다. 
- 노이즈, 공통분모 날리고 (약간의 에러와)주성분만으로 구성된 차원축소 결과물(핵심 알맹이만 남김)이다.

```python
(-2.68412563*w) 
```

3. 아래 예시 저차원데이터는 위 투영벡터를 저차원 벡터공간 기저벡터 기준으로 좌표변환한 것이다.
- 따라서 fit_transform()결과로 나온 저차원데이터는 '좌표변환 된 투영벡터'와 같은거다. 

```python
X1_low[0]
```

4. PCA 수행 결과로 주성분을 성공적으로 얻어냈다. 

```python
pca.components_
```

5. 한편 좌표변환 된 저차원 데이터는 차원이 달라서 원래 고차원 데이터와 유사도 비교가 안 된다. 
- 원래 데이터와 비교하기 위해 다시 '원점으로 중심을 옮기고', 고차원 기저벡터로 기준을 되돌리자.
- 아래 X2[0]은 위 조건이 적용된 고차원 데이터의 근삿값이다.

```python
X2[0] #저차원데이터를 다시 역변환한 뒤, 중심도 원점으로 완전히 되돌린상태
```

- 고차원데이터의 근삿값 X2[0] 뜯어보면 아래 [124] 처럼 구성되어 있다. 
- 핵심은 공통분모를 다시 더해서 온전한 근삿값을 만들었다는 점이다. 
- 또한 결과적으로 노이즈 날아간, 약간의 에러가 있지만 공통분모 + 주성분 만으로 구성된 근삿값 이라는 게 포인트다.

```python
pca.mean_+(-2.68412563*w) 
```

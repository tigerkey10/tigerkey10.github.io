# 부호함수 

```python
def f(x) : 
    if x > 0 : 
        return 1
    elif x == 0 : 
        return 0
    elif x < 0 : 
        return -1
xx = np.linspace(-2,2,100)

sign = [f(x) for x in xx]
plt.plot(xx, sign, 'b')
plt.title('부호함수')
plt.show()
```

# 단위계단 함수 

```python
def sidestep(x) : 
    if x >=0 : 
        return 1
    elif x < 0 : 
        return 0 
sidestep = [sidestep(x) for x in xx]
plt.subplot(211)
plt.plot(xx, sidestep, 'b')
plt.title('단위계단 함수')

plt.subplot(212)
plt.plot(xx, sign, 'r')
plt.title('부호함수')
plt.tight_layout()
plt.show()
```

# 지시함수 
### 내가 지정해준 값은 1 출력, 아니면 0 출력 

```python
def f(a, b) : 
    if a == b : 
        return 1
    elif a != b : 
        return 0 
list1 = [f(a,1) for a in xx]
print(list1, end= '')
```

# matplotlib으로 함수 시각화 하기 

```python
def f(x) : 
    return x**3-3*x**2+x
xx = np.linspace(-1,3,500)
yy = [f(a) for a in xx]

plt.plot(xx, yy, 'b')
plt.title('함수 시각화')
plt.show()
```

# 라인플롯은 x, y 받아서 벡터 생성, 각 점 찍고
# 각 벡터 점을 연결해서 그래프 만든다. 

```python
# 역함수 그래프 그리기 y=x 대칭인 함수 
def f1(x) : 
    return x**2
def f1inv(x) : 
    return np.sqrt(x)

x = np.linspace(0, 3, 300)
plt.plot(x, f1(x), 'r-', label='원래 함수')
plt.plot(x, f1inv(x), 'b', label='원래 함수의 역함수')
plt.plot(x, x, 'g--', label='$y=x$')
plt.axis('equal')
plt.xlim(0,3)
plt.ylim(0, 3)
plt.title('함수와 그 역함수')
plt.legend()
plt.show()
```

# 렐루 함수 
## 변형된 최대함수max()

```python
def relu(x) : 
    if x >=0 : 
        return x
    elif x < 0 : 
        return 0 
xx = np.linspace(-10,10, 100)
yy = [relu(x) for x in xx]
plt.plot(xx, yy, 'b')
plt.title('$ReLU$ 함수 or $max(x, 0)$')
plt.text(0,0, '0.0에서 불연속')
plt.show()
```

# 로지스틱 함수 _시그모이드 함수 
### 지수함수의 변형

```python
def logistic(x) : 
    return 1/(1+np.exp(-x))
xx = np.linspace(-10, 10, 100)
yy = [logistic(a) for a in xx]

plt.plot(xx, yy, 'b')
plt.title('로지스틱_시그모이드 함수')
plt.ylabel('$logistic(x)$')
plt.xlabel('$x$')
plt.show()
```

## 소프트맥스 함수 

```python
def softmax(x,w) : 
    e = np.exp(w*x)
    return np.exp(w*x)/e.sum()
x = [2.0, 1.0, 0.5]
y = softmax(x, np.ones(3))
```

```python
y
```

```python
np.sum(y)
```

# 소프트맥스 함수는 입력을 0과 1사이 확률값처럼 보이게 만들어준다. 

# 함수 이동시키기 

```python
xx = np.linspace(-10, 10, 100)
plt.plot(xx, logistic(xx-3)-2, label='$logistic(xx-3)-2$')
plt.plot(xx, logistic(xx), label='$logistic(x)$')
plt.title('로지스틱 함수 오른쪽으로 3, 아래쪽으로 2만큼 이동시키기')
plt.legend()
plt.show()
```

```python
def g(x,y) : 
    return np.exp(-x**2-16*y**2)

xx = np.linspace(-1,1,100)
yy = np.linspace(-1,1,100)
X, Y = np.meshgrid(xx,yy)
Z1 = g(X,Y)
Z2 = g(X-0.3,Y+0.7) #2차원 함수 오른쪽으로 0.3만큼, 아래쪽으로 0.3만큼 이동

plt.contour(X,Y,Z1)
plt.contour(X,Y,Z2, linestyles='--')
plt.text(0,0, 'g(x,y)')
plt.text(0.13, -0.7, 'g(x-0.3, y+0.7)')
plt.title('다변수함수의 평행이동')
plt.show()
```

# 함수 스케일링 
## 함수 축방향으로 잡아 늘리기 

```python
xx = np.linspace(-6,6,100)
plt.plot(xx, logistic(xx), label='$\sigma(x)$', ls='--')
plt.plot(xx, (1/3)*logistic(xx/2), label='(1/3)$\sigma(x/2)$', ls='-') #x축 방향으로 2배 늘리고. y축 방향으로 1/3배 늘린 것
plt.legend()
plt.title('로지스틱 함수를 x축 방향 2배 늘리고 , y축 방향으로 1/3배 늘린 것')
plt.show()
```

# 수치미분
## 0에 가까운 dx값 직접 주고 
## 대략적 기울기 구한다. 
## 사이파이.derivative() 명령 쓴다. 

```python
def f(x) : 
    return x**3-3*x**2+x
from scipy.misc import derivative

print(derivative(f, 0, dx=1e-6))
print(derivative(f, 1, dx=1e-6)) # 입력 위치는 x위치 들어간다. 
```

# 함수와 도함수 

```python
def f(x) : 
    return x**3-3*x**2+x
def fprime(x) : 
    return 3*x**2-6*x+1

x1,x2 = np.roots([3,-6,1])

xx = np.linspace(-1,3,600)
plt.subplot(211)
plt.plot(xx, f(xx))
plt.title('$f(x)=x^{3}-3x^{2}+x$')
plt.axvline(x1, c='b', ls='--')
plt.axvline(x2, c='b', ls='--')

plt.subplot(212)
plt.plot(xx, fprime(xx), 'r')
plt.title('$fprime(x)=3x^{2}-6x+1$')
plt.tight_layout()

plt.axvline(x1, c='b', ls='--')
plt.axvline(x2, c='b', ls='--')
plt.show()
```

# 심파이 이용한 심볼릭 연산 & 미분 
### sympy.diff(함수, 미분할 변수)

```python
import sympy
x = sympy.symbols('x')

f = x*sympy.exp(x)
f #심볼릭 변수 이용해 만든 함수 
```

```python
sympy.simplify(sympy.diff(f))
```

```python
x, y = sympy.symbols('x,y')
f = x**2+4*x*y+4*y**2
f
```

$x$편미분한 1차 도함수 

```python
sympy.simplify(sympy.diff(f, x)) 
```

$y$편미분한 1차 도함수 

```python
sympy.diff(f, y)
```

```python
x, mu, sigma = sympy.symbols('x, mu, sigma')
f = sympy.exp((x-mu)**2/sigma**2)
f
```

```python
sympy.simplify(sympy.diff(f, x))
```

이차 도함수 

```python
sympy.diff(f, x, x)
```

```python
# 연습문제 4.2.5
x = sympy.symbols('x')

#1. 
f_1 = x**3-1
sympy.diff(f_1,x)
```

```python
#2. 
k = sympy.symbols('k')
f_2 = sympy.log(x**2-3*k)
sympy.diff(f_2, x)
```

```python
#3. 
a, b, x = sympy.symbols('a, b, x')
f_3 = sympy.exp(a*x**b)
sympy.simplify(sympy.diff(f_3, x))
```

```python
#4.2.6 연습문제 
x, y = sympy.symbols('x,y')
f = sympy.exp(x**2+2*y**2)
f
```

```python
# x로 1차 편미분
sympy.diff(f,x)
```

```python
# y로 1차 편미분
sympy.diff(f, y)
```

```python
# x로 1차 편미분 한 도함수 x로 2차 편미분 --> 2차 도함수 
sympy.simplify(sympy.diff(f,x, x))
```

```python
# x로 1차 편미분 한 도함수 y로 2차 편미분 --> 2차 도함수 
sympy.simplify(sympy.diff(f,x, y))
```

```python
# y로 1차 편미분 한 도함수 x로 2차 편미분 --> 2차 도함수 
sympy.simplify(sympy.diff(f,y, x))
```

## '편미분 순서는 상관 없다'- 슈와르츠 정리 성립하는 걸 볼 수 있다. 

```python
# y로 1차 편미분 한 도함수 y로 2차 편미분 --> 2차 도함수 
sympy.simplify(sympy.diff(f,y,y))
```

# λ¶€νΈν•¨μ 

```python
def f(x) : 
    if x > 0 : 
        return 1
    elif x == 0 : 
        return 0
    elif x < 0 : 
        return -1
xx = np.linspace(-2,2,100)

sign = [f(x) for x in xx]
plt.plot(xx, sign, 'b')
plt.title('λ¶€νΈν•¨μ')
plt.show()
```

# λ‹¨μ„κ³„λ‹¨ ν•¨μ 

```python
def sidestep(x) : 
    if x >=0 : 
        return 1
    elif x < 0 : 
        return 0 
sidestep = [sidestep(x) for x in xx]
plt.subplot(211)
plt.plot(xx, sidestep, 'b')
plt.title('λ‹¨μ„κ³„λ‹¨ ν•¨μ')

plt.subplot(212)
plt.plot(xx, sign, 'r')
plt.title('λ¶€νΈν•¨μ')
plt.tight_layout()
plt.show()
```

# λΌμΈν”λ΅―μ€ x, y λ°›μ•„μ„ λ²΅ν„° μƒμ„±, κ° μ  μ°κ³ 
# κ° λ²΅ν„° μ μ„ μ—°κ²°ν•΄μ„ κ·Έλν”„ λ§λ“ λ‹¤. 

```python
# μ—­ν•¨μ κ·Έλν”„ κ·Έλ¦¬κΈ° y=x λ€μΉ­μΈ ν•¨μ 
def f1(x) : 
    return x**2
def f1inv(x) : 
    return np.sqrt(x)

x = np.linspace(0, 3, 300)
plt.plot(x, f1(x), 'r-', label='μ›λ ν•¨μ')
plt.plot(x, f1inv(x), 'b', label='μ›λ ν•¨μμ μ—­ν•¨μ')
plt.plot(x, x, 'g--', label='$y=x$')
plt.axis('equal')
plt.xlim(0,3)
plt.ylim(0, 3)
plt.title('ν•¨μμ™€ κ·Έ μ—­ν•¨μ')
plt.legend()
plt.show()
```

# λ΅μ§€μ¤ν‹± ν•¨μ _μ‹κ·Έλ¨μ΄λ“ ν•¨μ

```python
def logistic(x) : 
    return 1/(1+np.exp(-x))
xx = np.linspace(-10, 10, 100)
yy = [logistic(a) for a in xx]

plt.plot(xx, yy, 'b')
plt.title('λ΅μ§€μ¤ν‹±_μ‹κ·Έλ¨μ΄λ“ ν•¨μ')
plt.ylabel('$logistic(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
def softmax(x,w) : 
    e = np.exp(w*x)
    return np.exp(w*x)/e.sum()
x = [2.0, 1.0, 0.5]
y = softmax(x, np.ones(3))
```

# μ†ν”„νΈλ§¥μ¤ ν•¨μλ” μ…λ ¥μ„ 0κ³Ό 1μ‚¬μ΄ ν™•λ¥ κ°’μ²λΌ λ³΄μ΄κ² λ§λ“¤μ–΄μ¤€λ‹¤. 

# ν•¨μ μ΄λ™μ‹ν‚¤κΈ° 

```python
xx = np.linspace(-10, 10, 100)
plt.plot(xx, logistic(xx-3)-2, label='$logistic(xx-3)-2$')
plt.plot(xx, logistic(xx), label='$logistic(x)$')
plt.title('λ΅μ§€μ¤ν‹± ν•¨μ μ¤λ¥Έμ½μΌλ΅ 3, μ•„λμ½μΌλ΅ 2λ§νΌ μ΄λ™μ‹ν‚¤κΈ°')
plt.legend()
plt.show()
```

```python
def g(x,y) : 
    return np.exp(-x**2-16*y**2)

xx = np.linspace(-1,1,100)
yy = np.linspace(-1,1,100)
X, Y = np.meshgrid(xx,yy)
Z1 = g(X,Y)
Z2 = g(X-0.3,Y+0.7) #2μ°¨μ› ν•¨μ μ¤λ¥Έμ½μΌλ΅ 0.3λ§νΌ, μ•„λμ½μΌλ΅ 0.3λ§νΌ μ΄λ™

plt.contour(X,Y,Z1)
plt.contour(X,Y,Z2, linestyles='--')
plt.text(0,0, 'g(x,y)')
plt.text(0.13, -0.7, 'g(x-0.3, y+0.7)')
plt.title('λ‹¤λ³€μν•¨μμ ν‰ν–‰μ΄λ™')
plt.show()
```

# μμΉλ―Έλ¶„
## 0μ— κ°€κΉμ΄ dxκ°’ μ§μ ‘ μ£Όκ³  
## λ€λµμ  κΈ°μΈκΈ° κµ¬ν•λ‹¤. 
## μ‚¬μ΄νμ΄.derivative() λ…λ Ή μ“΄λ‹¤. 

```python
def f(x) : 
    return x**3-3*x**2+x
from scipy.misc import derivative

print(derivative(f, 0, dx=1e-6))
print(derivative(f, 1, dx=1e-6)) # μ…λ ¥ μ„μΉλ” xμ„μΉ λ“¤μ–΄κ°„λ‹¤. 
```

# ν•¨μμ™€ λ„ν•¨μ 

```python
def f(x) : 
    return x**3-3*x**2+x
def fprime(x) : 
    return 3*x**2-6*x+1

x1,x2 = np.roots([3,-6,1])

xx = np.linspace(-1,3,600)
plt.subplot(211)
plt.plot(xx, f(xx))
plt.title('$f(x)=x^{3}-3x^{2}+x$')
plt.axvline(x1, c='b', ls='--')
plt.axvline(x2, c='b', ls='--')

plt.subplot(212)
plt.plot(xx, fprime(xx), 'r')
plt.title('$fprime(x)=3x^{2}-6x+1$')
plt.tight_layout()

plt.axvline(x1, c='b', ls='--')
plt.axvline(x2, c='b', ls='--')
plt.show()
```

# μ‹¬νμ΄ μ΄μ©ν• μ‹¬λ³Όλ¦­ μ—°μ‚° & λ―Έλ¶„ 
### sympy.diff(ν•¨μ, λ―Έλ¶„ν•  λ³€μ)

```python
import sympy
x = sympy.symbols('x')

f = x*sympy.exp(x)
f #μ‹¬λ³Όλ¦­ λ³€μ μ΄μ©ν•΄ λ§λ“  ν•¨μ 
```

```python
sympy.simplify(sympy.diff(f))
```

```python
x, y = sympy.symbols('x,y')
f = x**2+4*x*y+4*y**2
f
```

```python
def F(b) : 
    if b <= 180 : 
        return (2/3)*(b/180)
    elif b >180 : 
        return (2/3)+((b-180)/540)
```

```python
xx = np.linspace(-200,500, 500)
yy = [F(i) for i in xx]
plt.plot(xx,yy)
```

```python
import sympy 
b = sympy.symbols('b')
f = (2/3)*(b/180)
sympy.diff(f)
```

```python
f2 = (2/3)+(b-180)/540
sympy.simplify(sympy.diff(f2))
1/540
```

```python
def p(b) : 
    if b <= 0 : 
        return 0 
    elif b <= 180 : 
        return 0.0037037037037037
    elif 180 < b and b <= 360 : 
        return 0.001851851851851852
    elif b > 360 : 
        return 0
```

```python
y2 = [p(i) for i in xx]
plt.plot(xx, xx*y2)
plt.xticks([-100,0,100,200,300,400,500])
plt.show()
```

```python
x = np.linspace(-100,500,1000)
p = np.zeros_like(x)
p[(0<x) & (x<=180)] = 2/(3*180)
p[(180<x) & (x<=360)] = 1/(3*180)
xp = x*p
```

```python
plt.plot(x,xp)
```

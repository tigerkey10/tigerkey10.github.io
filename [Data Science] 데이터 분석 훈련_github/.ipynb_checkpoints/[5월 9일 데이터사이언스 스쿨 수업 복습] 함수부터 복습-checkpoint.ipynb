# 단위계단 함수 & 부호함수 비교해보기 

```python
xx = np.linspace(-2, 2, 100)

def heavysidestep(x) : 
    if x >= 0 : 
        return 1
    elif x < 0 : 
        return 0
    
list1 = []
for i in xx : 
    result = heavysidestep(i)
    list1.append(result)
    
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(xx, list1)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('단위계단 함수(0에서 불연속)')
plt.show()

plt.subplot(212)
xx = np.linspace(-1, 1, 200)
yy = np.sign(xx)

plt.plot(xx, yy, )
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('부호함수(불연속)')
plt.show()
```

# 지시함수 

```python
def direct(x, direction) : 
    if x == direction : 
        return 1
    elif x != direction : 
        return 0 

xx = [1,2,1,2,4,5,3,1,2,3,]

list2 = []
for check in xx : 
    list2.append(direct(check, 1))

print(list2, end=' ')
plt.plot(xx, list2, 'ro')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('지시함수')
```

# 함수 시각화 하기 - matplotlib 라인플롯 
- x값들과 y값들 입력받아, '점' 생성하고, 좌표평면에 점 찍은 뒤 선으로 이어서 라인플롯 그린다. 

```python
def f(x) : 
    return x**3-3*x**2+x

#x값 벡터
xx = np.linspace(-5, 5, 8)

#y값 계산 --> y값 벡터 생성
yy = f(xx)

# 점 찍어서 함수 시각화 
plt.plot(xx, yy, )
plt.xlabel('$x$')
plt.ylabel('$f(x) = x^{3}-3x^{2}+x$')
plt.title('함수 시각화하기 연습')
plt.xticks(np.arange(-5, 6))
plt.show()
```

# ReLU 함수 
- 최대함수 max(x, y)에서 y가 0으로 고정된 특수형태 

```python
def ReLU(x) : 
    if x >= 0 : 
        return x
    elif x < 0 : 
        return 0
plt.figure()

xx = np.linspace(-5, 5, 200)
list1 = []
for i in xx : 
    result = ReLU(i)
    list1.append(result)

plt.plot(xx, list1, label='ReLU 함수')
plt.xlabel('$x$')
plt.ylabel('$ReLU(x)$')
plt.title('ReLU함수')
plt.legend()
plt.text(-1, 0, '원점에서 불연속(=미분 불가능)')
plt.show()
```

```python
def f(x) : 
    return x**3-12*x+20*np.sin(x)+7
xx = np.linspace(-3, 3, 100)
yy = f(xx)

plt.subplot(211)
plt.plot(xx, yy, label='log취하기 전')
plt.xticks([-1, 1], labels=['최저점', '최고점'])
plt.legend()
plt.axhline(0, c='r', ls='--')
plt.title('$\log$취하기 전')

plt.subplot(212)
plt.plot(xx, np.log(yy), label='log취한 후')
plt.xticks([-1, 1])
plt.legend()
plt.axhline(0, c='r', ls='--')
plt.title('$\log$취한 후')

plt.show()
```

# 함수에 log 취해도, 최고점 최저점 상대적 값은 변함 없다. 
- 절대적 값은 log취하면 더 강조된다. 
- log 취한 결과가 훨씬 눈에 잘 띄고, 가시성이 좋다. 
- 최적화 할 때 결과값에 주로 log취해서 본다. 

# 소프트맥스 함수 

```python
def softmax(x, w) : 
    e = np.exp(w*x)
    return np.exp(w*x)/e.sum()

x = np.array([2.0,1,0.5])
softmax(x, np.ones(3))
```

- 입력한 3차원 벡터의 값들에 가중치 반영, 확률값 비슷한 꼴들로 바뀌어 출력됬다. 

```python
def softmax(x, w) : 
    e = np.exp(w*x)
    return np.exp(w*x)/e.sum()

x = np.array([2.0,1,0.5])
softmax(x, 5*np.ones(3))
```

- 가중치를 높이니 최댓값과 최솟값 사이 거리가 더 벌어졌다(늘어졌다)

# 단변수함수 평행이동 

```python
xx = np.linspace(-10, 10, 100)
def logistic(x) : 
    return 1/(1+np.exp(-x))

plt.subplot(211)
plt.plot(xx, logistic(xx), 'r',ls='--', label='평행이동 전')

#오른쪽으로 5만큼, 위로 2만큼 평행이동 해보자. 
plt.plot(xx, logistic(xx-5)+2, 'b', label='평행이동 후')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)=logistic(x)$')
plt.title('단변수함수 평행이동')

#######
plt.subplot(212)
plt.title('부호함수 평행이동')
xx2 = np.linspace(-10, 10, 100)
yy2 = np.sign(xx2)
plt.xlabel('$x$')
plt.ylabel('$f(x)=sign(x)$')
plt.plot(xx2, yy2, 'r', ls='--', label='평행이동 전')
plt.plot(xx2, np.sign(xx2-5)+2, 'b', label='평행이동 후')
plt.legend()

plt.tight_layout()
plt.show()
```

# 다변수함수 이동


```python
fig = plt.figure()
ax = Axes3D(fig)
def g(x, y) : 
    return x**3+3*x**2-y**3+7*y**2-3*y-35
xx = np.arange(-10,10, 0.25)
yy = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(xx, yy)
Z = g(X,Y)
ax.plot_surface(X,Y,Z,rstride=1, cstride=1, cmap='hot')

Z2 = g(X-3, Y-5)
ax.plot_surface(X,Y,Z2, rstride=1, cstride=1, cmap='cool')
```

```python
def g(x, y) : 
    return np.exp(-x**2-16*y**2)

xx = np.linspace(-1,1, 100)
yy = np.linspace(-1,1, 100)

X,Y = np.meshgrid(xx, yy)
Z = g(X, Y)
Z2 = g(X-0.2, Y-0.35)

plt.contour(X,Y,Z, cmap='hot')
plt.contour(X,Y,Z2, cmap='cool')
plt.show()
```

# 다변수함수를 오른쪽으로 0.2, 위쪽으로 0.35 이동

# 단변수함수의 스케일링

```python
xx1 = np.linspace(-6, 6, 100)
def logistic(x) : 
    return 1/(1+np.exp(-x))
yy1 = logistic(xx1)

plt.plot(xx1, yy1,'r', label='$f(x)=logistic(x)$',)
plt.plot(xx1, (1/2)*(logistic((xx1+3)/2)+1), 'b', label='moved+scaled', ls='--')
plt.legend()
plt.title('로지스틱 함수 평행이동 + 스케일링')
plt.show()
```

# PCA 주성분 분석 복습

```python
from sklearn.decomposition import PCA
pca1 = PCA(n_components=1)

from sklearn.datasets import load_iris
data = load_iris()
X = data.data

#1차원 근사 
X_low = pca1.fit_transform(X)
X_low #기저에서 각 변수값들 결정짓는 단 하나의 공통요인 # 잠재변수 # 주성분
```

## 한편, inverse_transform한 결과는 평균 + 주성분으로 구성된 벡터와 같았다. 
## 투영벡터인데, 평균을 더해준 것이다.

## 이론적으로 주성분은 
## 특징행렬 X의 오른쪽 특이벡터 또는 $X^{T}X$분산행렬의 고유벡터와 같았다. 
## 같은지 확인해 보자. 
## 같다면, 가장 큰 특잇값의 제곱에 대응되는 오른쪽 첫번째 특이벡터 or 가장 큰 고윳값에 대응되는 고유벡터와 같을 것이다. 

```python
U, SG, VT = np.linalg.svd(X-pca1.mean_)
print(SG)
print(VT[0]) #오른쪽 첫번째 특이벡터와 일치한다. 
```

```python
pca1.components_
```

```python
X_ = X-pca1.mean_

ld, V = np.linalg.eig(X_.T@X_)
print(ld)
print(V[:,0]) # 가장 큰 고윳값에 대응되는 고유벡터와도 일치한다. 
```

# 올리베터 얼굴사진으로 이미지 PCA 

```python
from sklearn.datasets import fetch_olivetti_faces
faces_all = fetch_olivetti_faces()
K = 4
faces = faces_all.images[faces_all.target==K]

N = 2
M = 5 
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) : 
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(faces[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle('올리베티 얼굴사진')
plt.tight_layout()
plt.show()
```

# 이 2차원 배열 이미지 객체들을 주성분 2개로 분해해보자. 

# 주성분 : 공통요인으로 포함되어 각 데이터를 다르게 만드는 주요요인 

```python
from sklearn.decomposition import PCA
X3 = faces_all.data[faces_all.target==4]
pca3 = PCA(n_components=2)
X_low3 = pca3.fit_transform(X3)
X_low3 #주성분 값들 
```

# 주성분 만으로 구성된 투영벡터

```python
pca3.inverse_transform(X_low3)[0]-pca3.mean_
```

```python
(pca3.components_[0]*X_low3[0,0])+(pca3.components_[1]*X_low3[0,1])
```

# 주성분

```python
pca3.components_
```

```python
X32 = pca3.inverse_transform(X_low3)
```

```python
N = 2
M = 5 
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) : 
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(X32[k].reshape(64,64), cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle('노이즈 날리고 평균+주성분 만으로 구성된 근사데이터')
plt.tight_layout()
plt.show()
```

# 얼굴사진의 주요 특징(차이) 형성하는 각 주성분 얼굴 찾았다. 
# 이제 각 주성분들이 얼마씩 들어있느냐에 따라 얼굴사진 특징이 조금씩 달라질 것이다. 


# 각 주성분이 어떤 특징 형성하는지 알아보자. 

# 주성분 1

```python
N = 2
M = 5 
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) : 
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k<5 else 1.5*(k-4)
        ax.imshow((pca3.mean_+w*pca3.components_[0]).reshape(64,64) ,cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(f'평균+주성분{w}')
plt.suptitle('노이즈 날리고 평균+주성분 만으로 구성된 근사데이터')

plt.show()
```

# 이 주성분은 얼굴 방향을 결정짓는 주성분이다. 
# 이 주성분 +양의 방향으로 많아질 수록 인물이 정면을 본다. 
# 이 주성분 -음의 방향으로 많아질수록 인물 얼굴이 사선을 향하고 있다.

# 주성분 2

```python
N = 2
M = 5 
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) : 
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k<5 else 1.5*(k-4)
        ax.imshow((pca3.mean_+w*pca3.components_[1]).reshape(64,64) ,cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(f'평균+주성분{w}')
plt.suptitle('노이즈 날리고 평균+주성분 만으로 구성된 근사데이터')

plt.show()
```

# 연간 주가 수익률 데이터 PCA

```python
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import datetime

symbols = [
    'SPASTT01USM661N',
    'SPASTT01JPM661N',
    'SPASTT01EZM661N',
    'SPASTT01KRM661N'
]

data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred',
                             start=datetime.datetime(2000,1,1),
                             end=datetime.datetime(2020,12,31))[sym]
data.columns = ['US', 'JP', 'EZ', 'KR']
data = data/data.iloc[0]*100

styles = ['b-.', 'g--', 'c:','r-']
data.plot(style=styles)
plt.title('세계 주요국 20년간 주가')
plt.show()
```

```python
df = ((data.pct_change()+1).resample('A').prod()-1).T*100
print(df.iloc[:,:5])
```

```python
df.T.plot(style=styles)
plt.title('주요국 20년간 연간 수익률')
plt.xticks(df.columns)
plt.show()
```

# 위 그래프를 보면. 각 데이터 변이에 규칙성이 보인다. 
# 각 데이터 변이에 공통 주성분이 기저에서 작용하고 있다고 생각할 수 있다. 

# 그렇다면 이 주성분 찾고, 이 주성분은 어떤 주성분인가 알아보자.

## - 1개 레코드 기준. 각 변수가 일정 비율로. 규칙적 변이 하도록 만드는 요인 : 주성분
## - 또한 데이터를 평균수준에서 변화시키는 요인 : 주성분

## 여러 레코드에서 변수 변이 양상이 비슷하게. 규칙적이라면 : 공통 주성분이 있을 것이다! 라고 추측 가능

```python
pca4 = PCA(n_components=1)
w = pca4.fit_transform(df)
df2 = pd.DataFrame(w, 
            index=['US', 'JP','EZ','KR'])
df2.rename(mapper={0:'주성분비중'}, axis=1, inplace=True)
```

```python
df2
```

```python
m = pca4.mean_
m
```

```python
# 각 레코드 공통 주성분
p1 = pca4.components_
p1[0]
```

```python
# 평균과 주성분 만으로 구성된 근사데이터 
df_i = pd.DataFrame(pca4.inverse_transform(w))
df_i.index = df.index
df_i.columns = df.columns
df_i.iloc[:,-10:]
df_i.T.plot(style=styles)


plt.title('평균과 주성분 만으로 구성된 고차원 근사데이터')
plt.xticks(df.columns)
plt.show()
```

## 미국.유럽.일본 세 나라는 대체적으로 주성분이 비슷하게 들어있다. 
## 한국은 나머지 세 나라와 좀 다른 주성분 비중 을 가지고 있다. 

```python
df2
```

## 주성분이 
### 한국은 약 + 72만큼 들어있다. 
### 유럽.일본.미국은 모두 이 주성분 비중이 -수치다. 

한국은 + 쪽으로 많이 들어있고, 나머지 세 나라는 - 쪽으로 많이 들어있는 이 주성분은 뭘까? 

# 주성분이 + 부호로 추가될 수록 

```python
x_range = np.linspace(2000,2020, 21, dtype=int)
for i in np.linspace(0, 100, 5) : 
    plt.plot(x_range, (pca4.mean_+p1*i).T)
    
plt.plot(x_range, (pca4.mean_+p1*100).T, label='주성분 100배 추가한 수익률')
plt.plot(x_range, pca4.mean_.T, 'ko-', lw=5, label='평균 수익률')
plt.title('주성분 비중 변화에 따른 주가 수익률 변화')
plt.legend()
plt.show()
```

## 주성분 비중 + 양 부호로 추가될 수록 
## 평균수준에서 위쪽으로 변화한다. 

# 주성분이 - 방향 추가될 수록 

```python
x_range = np.linspace(2000,2020, 21, dtype=int)
for i in np.linspace(-100,0, 5) : 
    plt.plot(x_range, (pca4.mean_+p1*i).T)
    
plt.plot(x_range, (pca4.mean_+p1*-100).T, label='주성분 100배 추가한 수익률')
plt.plot(x_range, pca4.mean_.T, 'ko-', lw=5, label='평균 수익률')
plt.title('주성분 비중 변화에 따른 주가 수익률 변화')
plt.legend()
plt.show()
```

# 주성분이 - 방향 추가 될 수록 데이터가 위 그래프 처럼 변한다. 

# 한국은 주성분 비중이 + 였고 나머지 세 국가는 - 였다. 
## 한국 연간 수익률 그래프는 첫번째 그래프 처럼. 평균수준에서 위쪽으로 변화할 것이다. 
## 나머지 3국 연간 수익률 그래프는 두번째 그래프 처럼 평균 수준에서 아래쪽으로 변화할 것이다. 

## 한국에는 + 쪽으로 작용하고, 나머지 3국에는 - 쪽으로 작용하는 주성분.
## 한국을 선진국 3개 국 그래프에서 멀어지게 만드는 주성분.
## + 작용할 때 선진국 3개 그래프에서 한국 그래프 쪽으로 변화시키는 주성분

# 중진국 성분이다. 

# 기본 전제 : 여러 데이터가 있다. 
# 여러 데이터 간 변이 양상이 규칙성을 띈다. --> 공통 주성분이 작용하기 때문이다. 
## 주성분 분석을 시도해 볼 수 있다. 

# 데이터 평균수준에 기반 놓고, 거기서 각 데이터 간 차이를 만들어 내는 것이 주성분이다. 
## 이 차이에도 규칙성이 존재한다.

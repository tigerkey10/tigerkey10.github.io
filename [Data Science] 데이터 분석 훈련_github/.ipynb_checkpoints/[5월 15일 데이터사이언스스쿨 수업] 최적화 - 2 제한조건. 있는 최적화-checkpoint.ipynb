```python
def f1(x1, x2) : 
    return x1**2+x2**2

x1 = np.linspace(-5,5,100)
x2 = np.linspace(-3,3,100)
X1,X2 = np.meshgrid(x1,x2)
Z = f1(X1,X2)

x2_g = 1-x1

plt.contour(X1,X2,Z, colors='grey', levels=[0.5,2,8,32])
plt.plot(x1, x2_g, 'g-')

plt.plot([0],[0], 'rP')
plt.plot([0.5],[0.5], 'ro', ms=10)

plt.xlim(-5,5)
plt.ylim(-3,3)
plt.xticks(np.linspace(-4,4,9))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('등식 제한조건 있는 최적화 문제')
plt.show()
```

# fmin_slsqp 명령 : 사이파이로 제한조건 있는 최적화 하기 
- 뉴턴방법 최적화랑 방법은 같다(목적함수, 초깃값)
- 하지만 최적해가 제한조건도 동시에 만족하도록 조건 걸어준다. 

```python
def f1array(x) : 
    return x[0]**2+x[1]**2

def eq_constraint(x) : 
    return x[0]+x[1]-1

sp.optimize.fmin_slsqp(f1array, (10,11), eqcons=[eq_constraint, ])
```

# 부등식 제한조건 있는 최적화 문제 해결하기

```python
def f2(x) : 
    return np.sqrt((x[0]-4)**2)

k = 1
def ieq_constraint2(x) : 
    return np.atleast_1d(-np.abs(x[0])-np.abs(x[1])+k)

result = sp.optimize.fmin_slsqp(f2, np.array([0,0]), ieqcons=[ieq_constraint2])
```

```python
#5.2.4 연습문제 
def f2(x) : 
    return np.sqrt((x[0]-4)**2+(x[1]-2)**2)

k = float(input('k값 입력 : '))
def ieq_constraint2(x) : 
    return np.atleast_1d(-np.abs(x[0])-np.abs(x[1])+k)

optimized = sp.optimize.fmin_slsqp(f2, np.array([0,0]), ieqcons=[ieq_constraint2]) 
# 부등식 제한조건에 따라 최적화된 
# 함수 f2의 최적해 

##################


def f2(x, y) : 
    return np.sqrt((x-4)**2+(y-2)**2)

x1 = np.linspace(-2, 5, 100)
x2 = np.linspace(-1.5, 3, 100)
X1, X2 = np.meshgrid(x1,x2)
Y = f2(X1,X2)

plt.contour(X1, X2, Y, colors='gray', 
           levels=np.arange(0.5, 5, 0.5)*np.sqrt(2))
#############_ 등고선 그래프로 2차원 함수 f2 표현

ax = plt.gca()
x12 = np.linspace(-k, 0, 10)
x13 = np.linspace(0, k, 10)
ax.fill_between(x12, x12+k, -k-x12, color='g', alpha=0.5)
ax.fill_between(x13, x13-k, k-x13, color='g', alpha=0.5) #부등식 제한조건 영역 표시 


##최적점 위치 
x1_sol = optimized[0]
x2_sol = optimized[1]
plt.plot(x1_sol, x2_sol, 'ro', ms=10)

plt.xlim(-2, 5)
plt.ylim(-1.5, 3)
plt.xticks(np.linspace(-2, 5, 8))
plt.yticks(np.linspace(-1, 3, 5))
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```

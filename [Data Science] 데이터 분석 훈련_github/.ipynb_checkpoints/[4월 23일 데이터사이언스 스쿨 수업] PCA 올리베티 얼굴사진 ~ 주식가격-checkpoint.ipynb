# 올리베티 얼굴사진 PCA


```python
from sklearn.datasets import fetch_olivetti_faces
faces_all = fetch_olivetti_faces()
K = 20 
faces = faces_all.images[faces_all.target == K]

from sklearn.decomposition import PCA
pca3 = PCA(n_components=2)
X3 = faces_all.data[faces_all.target == K]
w3 = pca3.fit_transform(X3)
```

```python
w3 #2차원 근사된 저차원 데이터 --> 각 원소 하나씩을 모두 주성분이라 한다
```

```python
pca3.components_ #내가 투영한 2차원 벡터공간 기저벡터
```

```python
face_p1 = pca3.components_[0].reshape(64, 64)
face_p2 = pca3.components_[1].reshape(64,64)

print(face_p1)
```

```python
face_mean = pca3.mean_.reshape(64, 64)
```

```python
#테스트
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.95)
plt.imshow(face_mean+ 3*face_p1+ 1*face_p2, cmap=plt.cm.bone) 
```

```python
N = 2
M = 5
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) : 
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k < 5 else 1.5 * (k-4)
        ax.imshow(face_mean + w * face_p1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(1,5))
plt.imshow(face_mean-6.0*face_p1, cmap=plt.cm.bone)

plt.axis('off')
plt.show()
```

```python
X32 = pca3.inverse_transform(w3)
plt.imshow(X32[7].reshape(64, 64)+X32[9].reshape(64,64), plt.cm.bone)
plt.axis('off')
```

```python
#3.5.2 연습문제
from sklearn.datasets import fetch_olivetti_faces
faces_all = fetch_olivetti_faces()
K = 25

faces = faces_all.images[faces_all.target ==K]
X = faces_all.data[faces_all.target ==K]

N = 2
M = 5
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) : 
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N , M, k+1)
        ax.imshow(faces[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle('비틀스 얼굴사진')
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca5 = PCA(n_components=2)
X_low = pca5.fit_transform(X)

X_re = pca5.inverse_transform(X_low)
```

```python
principal1 = pca5.components_[0].reshape(64,64)
principal2 = pca5.components_[1].reshape(64,64) #이미지의 두 가지 주성분(두 기저벡터)
data_mean = pca5.mean_ #데이터 평균 x0
```

```python
plt.figure(figsize=(1,5))
plt.imshow(data_mean.reshape(64,64), plt.cm.bone)

plt.title('평균얼굴')
plt.axis('off')
plt.show() 
```

```python
faces_mean = data_mean.reshape(64,64)

N = 2
M = 5

fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) :
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k < 5 else 1.5*(k-4)
        ax.imshow(faces_mean+w*principal1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title('주성분1 비중 %s' %w)
plt.suptitle('평균 얼굴에 주성분 1 더한 사진')
plt.tight_layout()
plt.show()
```

# 시선 정면을 보느냐, 오른쪽 아래를 보느냐 차이 나타낸다

```python
faces_mean = data_mean.reshape(64,64)

N = 2
M = 5

fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N) :
    for j in range(M) : 
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5*(k-5) if k < 5 else 1.5*(k-4)
        ax.imshow(faces_mean+w*principal2, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title('주성분2 비중 %s' %w)
plt.suptitle('평균 얼굴에 주성분 2 더한 사진')
plt.tight_layout()
plt.show()
```

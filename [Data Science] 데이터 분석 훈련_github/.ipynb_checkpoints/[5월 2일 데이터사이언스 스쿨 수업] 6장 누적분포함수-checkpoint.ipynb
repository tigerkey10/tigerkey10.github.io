```python
t = np.linspace(-100, 500, 100) # -100부터 500까지 잘게 쪼개진 100개 숫자 
F = t/360 #각 구간 확률값
F[t<0] = 0 #각 구간 확률값 중 0 넘어서 음의 무한대 쪽인 것들은 모두 0으로 처리 
F[t>360] = 1  # 각 구간 확률값 중 360넘어서 양의 무한대 쪽인 것들 모두 1로 처리 
plt.plot(t, F) # 각 구간 지점에서의 확률 
plt.ylim(-0.1, 1.1) #그래프 y축 구간
plt.xticks([0, 180, 360]) #그래프 x 축에 막대 세우기 
plt.title('누적분포함수') # 그래프 제목 
plt.xlabel('$x$ (도)') #x축에 라벨 붙이기 
plt.ylabel('$F(x)$') #y축에 라벨 붙이기 
plt.show()
```

```python
#6.4.4 연습문제 
# 누적분포함수 F(x)
def F(x) : 
    if x <= 0 : 
        P = 0
    if 0 < x <= 180 : 
        return (2/3)*(x/180)
    if 180 < x <= 360 : 
        return (2/3) + ((1/3)*((x-180)/180))
    if 360 < x : 
        return 1
F(270)
```

```python
P_list = []

for x in np.linspace(-100, 500, 1000) : 
    if x <= 0 : 
        P = 0 
        P_list.append(P)
    if 0 < x <= 180 : 
        P = (2/3)*(x/180)
        P_list.append(P)
    if 180 < x <= 360 : 
        P = (2/3) + ((1/3)*((x-180)/180))
        P_list.append(P)
    if 360 < x : 
        P = 1
        P_list.append(P)
```

```python
plt.plot(np.linspace(-100, 500, 1000), P_list) # 각 구간 지점에서의 확률 
plt.ylim(-0.1, 1.1) #그래프 y축 구간
plt.xticks([0, 180, 360]) #그래프 x 축에 막대 세우기 
plt.title('조작된 원판의 누적분포함수') # 그래프 제목 
plt.xlabel('$x$ (도)') #x축에 라벨 붙이기 
plt.ylabel('$F(x)$') #y축에 라벨 붙이기 
plt.show()
```

```python
#6.4.5 연습문제 
# 확률밀도함수 p(x)
t = np.linspace(-100, 500, 1000) # -100부터 500까지 구간을 1000개로 잘게 쪼게겠다
p = np.gradient(P_list, 600/1000)
plt.plot(t, p)
plt.ylim(-0.0001, p.max()*1.1)
plt.xticks([0, 180, 360]) #x 축에 0, 180, 360 쪽에 막대 세워라
plt.title('조작된 원반의 확률밀도함수')
plt.xlabel('$x$도')
plt.ylabel('$p(x)$')
plt.show()
```

```python
#6.4.5 연습문제 
def p(x) : 
    if x <= 0 : 
        return 0
    if 0 < x <= 180 : 
        return (2/3)/180 #각 점의 확률 #넓이 / 밑변 길이  = 높이 (=각 미세구간 점에서의 확률값)
    if 180 < x <= 360 : 
        return (1/3)/180 #넓이 / 밑변 길이 = 높이 (= 각 미세구간 점에서의 확률 값)
    if 360 < x : 
        return 0
p(181)
```

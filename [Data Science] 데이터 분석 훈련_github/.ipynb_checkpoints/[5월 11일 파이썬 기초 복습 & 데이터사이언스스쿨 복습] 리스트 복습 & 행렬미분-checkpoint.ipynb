딕셔너리 

```python
#01.
{}.append('1')# 딕셔너리는 append()로 추가 못한다. 
```

```python
#01. 
dic = {}
for i in range(1, 5) : 
    fruit = input('%s 번째로 좋아하는 과일 이름은 ? : ' %i)
    dic[i] = fruit
print(dic)
```

```python
#02. 
hate = input('싫어하는 과일 이름 입력하세요 : ')
dic.pop(hate)
print(dic)
```

# 딕셔너리에서 삭제할 때는 pop()을 사용하고, key로 지워야 한다. 

```python
#02. 
print(dic)
hate = int(input('싫어하는 과일 번호를 입력하세요 : '))
dic.pop(hate)
```

```python
#03. 딕셔너리 값이나 키로 정렬할 때는 sorted()를 쓴다. 
sorted(dic.values())
sorted(dic.keys())
```

# 행렬미분 복습 

그레디언트 벡터의 의미 

```python
xx = np.linspace(-5, 5, 100)
yy = np.linspace(-5, 5, 100)
def f(x, y) : 
    return 2*x**2+6*x*y+7*y**2-26*x-54*y+107
X, Y = np.meshgrid(xx, yy)
Z = f(X, Y)

ax = plt.gca(projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0.1)
ax.view_init(0,110)
```

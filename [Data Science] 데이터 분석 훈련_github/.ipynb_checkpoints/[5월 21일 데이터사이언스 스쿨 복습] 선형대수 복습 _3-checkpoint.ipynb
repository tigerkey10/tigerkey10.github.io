## P.202 예제 _ 정방_대칭행렬을 랭크-1 행렬 합으로 표현하기 

$A=\sum^{N}_{i=1}\lambda_{i}v_{i}v_{i}^{T}$

```python
A = np.array([[60,30,20],
             [30,20,15],
             [20,15,12]])

ld, V = np.linalg.eig(A)
print(ld) #고윳값
print(V) #고유벡터 
print()

v1 = V[:,0].reshape(3,1)
v2 = V[:,1].reshape(3,1)
v3 = V[:,2].reshape(3,1) # 각 고유벡터 

ld1 = ld[0]
ld2 = ld[1]
ld3 = ld[2]

# 랭크 -1행렬 합
print((ld1*v1@v1.T)+(ld2*v2@v2.T)+(ld3*v3@v3.T))
print()
print(A)
```

## 이번에는 $A^{-1}$을 랭크-1행렬 합으로 표현해 보자. 

```python
np.linalg.inv(A)
```

```python
ld_1 = 1/ld[0]
ld_2 = 1/ld[1]
ld_3 = 1/ld[2]

(ld_1*v1@v1.T)+(ld_2*v2@v2.T)+(ld_3*v3@v3.T)
```

## 결론 : 특이분해 해서 얻은 결과를 고유분해 해서 얻은 결과와 대응시킬 수 있다.
- 특잇값 제곱 (과 0)은 고윳값과 같다. 
- 행렬 $A^{T}A$ 고유벡터는 행렬 A 오른쪽 특이벡터와 같다. 

```python
# 연습문제 3.4.3
A = np.array([[3,-1],
             [1,3],
             [1,1]])
ld, V = np.linalg.eig(A@A.T)
idx = np.argsort(ld**2)[::-1]
ld = ld[idx]
V = V[:,idx]
print(ld)
print(V)
print()
LD = np.diag(ld)
LD
```

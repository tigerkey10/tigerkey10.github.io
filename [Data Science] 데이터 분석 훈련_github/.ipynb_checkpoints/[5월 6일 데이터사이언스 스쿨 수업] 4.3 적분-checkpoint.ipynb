# 심파이를 이용한 부정적분

```python
import sympy

sympy.init_printing(use_latex='mathjax')

x = sympy.symbols('x') #x를 심볼릭 연산 위한 심볼변수로 지정
f = x*sympy.exp(x)+sympy.exp(x)
f
```

```python
# 부정적분
sympy.integrate(f)
```

```python
# 다시 미분해보자. 
sympy.diff(sympy.integrate(f))
```

```python
from matplotlib.patches import Polygon #다각형인듯

def f(x) : 
    return x**3-3*x**2+x+6

a, b = 0, 2
x = np.linspace(a-0.5, b+0.5, 50)
y = f(x)

ax = plt.subplot(111)
plt.title('정적분의 예 : 구간면적')
plt.plot(x, y, 'r', linewidth=2)

plt.ylim(bottom=0)
ix = np.linspace(a, b)
iy = f(ix)
verts = [(a, 0)] + list(zip(ix, iy))+[(b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5') #회색 다각형 
ax.add_patch(poly) #회색 다각형 그래프에 추가 
plt.text(0.5*(a+b), 0.2*(f(a)+f(b)), r'$\int_a^b f(x)dx$', #텍스트 찍을 좌표, 내용, 정렬=중앙, 폰트 사이즈 20 
        horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.05, '$x$')
plt.figtext(0.1, 0.9, '$y$')

ax.spines['right'].set_visible(False) #오른쪽 뼈대 안 보이게 
ax.spines['top'].set_visible(False) #가장 위쪽 뼈대(경계) 안 보이게 
ax.xaxis.set_ticks_position('bottom') #x축 가장 아래쪽에 구분막대 세우기 
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$')) #x축 아래 구분막대에 라벨 붙이기 
ax.set_yticks([])# y축 막대들 제거
ax.set_xlim(-2, 4)
ax.set_ylim(0, 8)

plt.show()
```

# 단변수함수의 정적분 구하기 
1. 미적분학의 기본정리 
2. 수치적분(대략적 면적 - 근사면적)

```python
x = sympy.symbols('x')
f = x**3-3*x**2+x+6
f
```

```python
# 미적분학 기본정리로 풀기 
F = sympy.integrate(f)
F
```

# 심볼릭 함수에 실제 숫자 넣어서 계산하려면 : 
- subs(), evalf() 메서드 쓴다. 
- subs()는 심볼릭 함수에 실제 숫자 넣고 계산한 결과 보여주고, 
- evalf()는 계산 마무리 된 결과 깔끔하게 정리해준다. 

```python
(F.subs(x, 2) - F.subs(x, 0))
```

```python
(F.subs(x, 2) - F.subs(x, 0)).evalf()
```

```python
# 연습문제 4.3.5
import sympy as sp

#1. 
x = sp.symbols('x')
def f1(x) : 
    return 3*x**2-6*x+1

F1 = sp.integrate(3*x**2-6*x+1) #원래 함수

#부정적분 이용해서 정적분 구하기 
(F1.subs(x, 1) - F1.subs(x, 0)).evalf()
```

```python
# 수치적분으로 구하기 (대략적 면적 - 근사면적)
import scipy 
scipy.integrate.quad(f1, 0, 1) #quad는 심파이 아니고 사이파이에 있다. 첫번째 argument에는 함수가 들어가야 한다. 
```

```python
#2. 
x = sympy.symbols('x')
def f2(x) : 
    return 2+6*x+4*sympy.exp(x)+5/x #프로그래밍 함수

f2 = 2+6*x+4*sympy.exp(x)+5/x #심볼릭 변수로 구성된 심볼릭 함수
f2
```

```python
F2 = sympy.integrate(f2)
F2 #부정적분 _ 심볼릭 함수
```

```python
((F2.subs(x, 10) - F2.subs(x, 1)))
```

```python
((F2.subs(x, 10) - F2.subs(x, 1))).evalf()
```

# 다변수 정적분

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
_x = np.arange(12)/2+2 #x 2에서 8까지 
_y = np.arange(12)/2 #y 0에서 6까지 범위
X, Y = np.meshgrid(_x, _y)
x, y = X.ravel(), Y.ravel() #ravel() : 1차원 배열(열벡터)로 전환
z = x*x-10*x+y+50
z0 = np.zeros_like(z) #z랑 크기 똑같은 영벡터 생성




ax.bar3d(x, y, z0, 0.48, 0.48, z) # x값, y값, z=0인 좌표에 가로세로 0.48씩, 높이는 z인 bar plot 생성
ax.set_xlim(0, 10)
ax.set_ylim(-2, 10)
ax.set_zlim(0, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.title('f(x, y)')
plt.show()
```

```python
x, y = sympy.symbols('x, y')

f_test = x**2-10*x+y+50
F_test = sympy.integrate(f_test, x)
F_mid = F_test.subs(x, 8) - F_test.subs(x, 2)
F_result = sympy.integrate(F_mid, y)

(F_result.subs(y, 6) - F_result.subs(y, 0)).evalf()
```

* 다변수함수 이중적분 할 때 x로 먼저 하든 y로 먼저하든 순서 상관 없다. 
- 슈와르츠 정리 때문이다. 

```python
f_test = x**2-10*x+y+50
F_test = sympy.integrate(f_test, y)
F_mid = F_test.subs(y, 6) - F_test.subs(y, 0)
F_result = sympy.integrate(F_mid, x)

(F_result.subs(x, 8) - F_result.subs(x, 2)).evalf()
```

# 슈와르츠 정리에 따르면 각각 다른 입력변수로 편미분 할 때 편미분 순서는 상관없었다. 

# 따라서 부정적분 or 정적분 할 때도 적분하는 입력변수 순서는 상관없다. 



```python
x,y = sympy.symbols('x,y')
test_f = x*y*sympy.exp(x**2+y**2)

sympy.integrate(sympy.integrate(test_f, y), x)
```

```python
x,y = sympy.symbols('x,y')
test_f = x*y*sympy.exp(x**2+y**2)

sympy.integrate(sympy.integrate(test_f, x), y)
```

# 람다함수 

- 함수 생성기 : lambda
- lambda 함수에 넣을 <인자> : 표현식 <함수 내부 표현식>
- lambda 인자 : 표현식
- def, 함수 이름이 lambda 안에 합쳐진거라 생각하면 된다. 

```python
y = (lambda x,y : 
     x+y)
y(10,20)
```

```python
# 예제 : 수치이중적분
def f(x, y) : 
    return np.exp(-x*y)/y**2

def x1(x) : 
    return 0
def x2(x) : 
    return np.inf

sp.integrate.dblquad(f, 1, np.inf, x1, x2) # y하한과 상한 의미하는 gfun과 hfun은 인자를 x로 갖는 함수 꼴 이어야 한다. 
```

```python
sp.integrate.dblquad(f, 1, np.inf, lambda x : 0 , lambda x : np.inf)
```

```python
#4.3.6 연습문제 
#수치이중적분
def f2(x, y) : 
    return 1+x*y

sp.integrate.dblquad(f2, -1, 1, lambda x : -1, lambda x : 1)
```

# 다차원 함수의 단일 정적분
- 규칙은 입력변수 1개에 대해 편미분 할 때랑 같다. 

```python
x, y = sympy.symbols('x,y')
f = 4*x**2+4*x*y+y**2

(sympy.integrate(f, x).subs(x, 7) - sympy.integrate(f, x).subs(x, 3)).evalf()
```

- y는 상수 취급

```python
from sklearn.datasets import load_iris
iris = load_iris()
N = 10
X = iris.data[:N, :2]

from sklearn.decomposition import PCA

pca1 = PCA(n_components=1) #주성분 벡터 갯수 설정 --> 1차원 벡터공간에 투영하겠다는 뜻이다
X_low = pca1.fit_transform(X) #2차원 벡터를 모두 1차원에 근사
X2 = pca1.inverse_transform(X_low) #1차원 근사한 결과를 다시 2차원으로 복귀 
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
X = load_iris()
X = X.data[:50,0:2]

pca1 = PCA(n_components=1) #1차원 근사하겠다
X_low = pca1.fit_transform(X) #1차원 근사 결과 
```

```python
X2 = pca1.inverse_transform(X_low)
X2[7] #2차원 상에서 X의 좌표 
```

```python
X[7] #원래 데이터랑 거의 비슷하지만 엄연히 다르다
```

```python
pca1.mean_ #PCA분석에서 조정된 원점 , 특징행렬 열별 평균
```

```python
pca1.components_ #1차원 근사에 사용한, 점과 직선사이 거리 가장 작아지는 기저벡터
```

```python
X0 = X - pca1.mean_
U, SG, VT = np.linalg.svd(X0)
print(SG)
print(VT)
VT[0]
```

```python
LD, V = np.linalg.eig(X0.T@X0)

print(LD)
print(V)
V[:, np.argmax(LD)]
```

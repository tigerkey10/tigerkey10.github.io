```python
# 10.1.1 연습문제 
pp = np.linspace(0, 1, 10000)
eps = np.finfo(float).eps
def H(x) : 
    return -(1-x)*np.log2(1-x)-(x*np.log2(x))
H = H(pp)
H[0] = 0
H[-1] = 0

plt.plot(pp, H)
plt.xlabel('$P(Y=1)$')
plt.ylabel('$H[Y]$')
plt.title('엔트로피 범함수 H[Y]')
plt.show()
```

```python
def H(x) : 
    if x == 0 or x == 1 : 
        return 0 
    else : 
        return -(1-x)*np.log2(1-x)-(x*np.log2(x))
r = [H(x) for x in pp]

plt.plot(pp, r)
```

# μ§€μ‹ν•¨μ 

```python
def direct(x, direction) : 
    if x == direction : 
        return 1
    elif x != direction : 
        return 0 

xx = [1,2,1,2,4,5,3,1,2,3,]

list2 = []
for check in xx : 
    list2.append(direct(check, 1))

print(list2, end=' ')
plt.plot(xx, list2, 'ro')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('μ§€μ‹ν•¨μ')
```

# ReLU ν•¨μ 
- μµλ€ν•¨μ max(x, y)μ—μ„ yκ°€ 0μΌλ΅ κ³ μ •λ νΉμν•νƒ 

```python
def ReLU(x) : 
    if x >= 0 : 
        return x
    elif x < 0 : 
        return 0
plt.figure()

xx = np.linspace(-5, 5, 200)
list1 = []
for i in xx : 
    result = ReLU(i)
    list1.append(result)

plt.plot(xx, list1, label='ReLU ν•¨μ')
plt.xlabel('$x$')
plt.ylabel('$ReLU(x)$')
plt.title('ReLUν•¨μ')
plt.legend()
plt.text(-1, 0, 'μ›μ μ—μ„ λ¶μ—°μ†(=λ―Έλ¶„ λ¶κ°€λ¥)')
plt.show()
```

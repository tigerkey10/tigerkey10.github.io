# 연습문제 2.2.5의 의의 : 
## 평균 제거벡터 만들기 위한 것이다. 
## X-X_ = 평균제거벡터 
### 주성분분석에 사용 
## $X^{bar}=$$(1/N)1_{N}1_{N}^{T}X$
## N은 행 갯수

```python
from sklearn.datasets import load_iris
X_ = load_iris()
X = X_.data[:10, 0:2]
X #입력 데이터 행렬
```

```python
n = np.ones(10).reshape(10,1)
X_ = (1/10)*(n@n.T@X)
X_ #평균행렬 
```

주성분 분석에 쓸. 평균제거행렬

```python
X-X_
```

```python
#2.4.5 연습문제 
from sklearn.datasets import load_boston
boston = load_boston()

X = boston.data
y = boston.target

A = X[:4,[0,4,5,6]]
B = y[:4]
```

$Ax=b$

```python
#1. 
x = np.linalg.inv(A)@B # 선형연립방정식의 미지수 해 = 선형회귀모형 
x # 직관과 일치 안 한다. 공해.범죄율은 거의 뭐 신경도 안 쓴다. 
```

```python
#2. 
x, resid, rank, s = np.linalg.lstsq(A, B)
x
```

```python
print(resid) #오차 없음
print(rank) #특징행렬 A는 풀랭크 - 각 열이 모두 선형독립 --> 역행렬도 존재
print(s) #행렬 A의 특잇값
```

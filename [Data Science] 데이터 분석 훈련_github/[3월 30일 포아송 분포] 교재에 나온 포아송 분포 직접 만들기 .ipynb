```python
def B(n,p) : 
    for x in range(0,(n+1)) : 
        comb = factorial(n)/(factorial(n-x)*factorial(x))
        p1 = (p**x)
        p2 = ((1-p)**(n-x))
        p_x = comb*p1*p2
        x_list.append(p_x)
    return x_list #이항분포 식
```

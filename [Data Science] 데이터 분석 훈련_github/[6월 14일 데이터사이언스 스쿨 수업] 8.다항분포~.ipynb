# 사이파이 이용한 다항분포 시뮬레이션

```python
N = 30
mu = [0.1]*4+[0.3]*2
rv = sp.stats.multinomial(30,mu) # 카테고리 확률변수 시뮬레이션 30번 후 나온 표본들 합 = rv 표본 1개 #
# 1개 표본 = N번 시행 중 각 주사위 눈금 x가 나온 빈도

# 이제 이 표본들의 확률분포를 알아보자. 
# 표본 시뮬레이션 100번
X = rv.rvs(100, random_state=0)

a = 3
print(len([i for i in X[:,0] if i == a]))
print([i for i in X[:,0] if i == a])
```

```python
df = pd.DataFrame(X)
df.index = np.arange(1,101)
df.columns=np.arange(1,7)
df2 = df.stack().reset_index()
df2.columns=['시행','카테고리값','표본값']

plt.figure(figsize=(10,5))
sns.violinplot(x='카테고리값',y='표본값',data=df2,inner='quartile')
sns.swarmplot(x='카테고리값',y='표본값',data=df2,color='.3')
plt.title('다항분포의 시뮬레이션결과')
plt.show()
```

```python
np.random.seed(15)
N = 7
data = np.sort(np.random.normal(size=(N)))[::-1]

sample_mean = np.mean(data)
print('한쪽에 치우쳐진 표본의 표본평균 : {}'.format(sample_mean))
print()

plt.figure(figsize=(10,2))
sns.rugplot(data, height=0.5, linewidth=4)
x = np.linspace(-3,3,100)
plt.axvline(x=0, ls=':',c='r',linewidth=2,label='실제 확률변수 기댓값')
plt.axvline(x=sample_mean, ls='--',c='g',linewidth=2, label='표본평균')
plt.legend()
plt.xlim(-2,2)
plt.title('표본평균이 데이터가 몰려있는 쪽에 편향되어 있다')
plt.show()
```

```python
# 함수의 기울기 
x = np.linspace(-1, 3, 400)
def f(x) : 
    return x**3-3*x**2+x

plt.plot(x, f(x))
plt.plot(0,0, 'ro')
plt.plot(x,x,'r:')
plt.plot(1, -1, 'go')
plt.plot(x, (3*1**2-6*1+1)*(x-1)-1, 'g--')

plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))

plt.annotate('', xy=[1,0], xytext=[0,0], arrowprops={'facecolor':'black'})
plt.annotate('', xy=[1,1], xytext=[1,0], arrowprops={'facecolor':'black'})

plt.annotate('', xy=[2,-1], xytext=[1,-1], arrowprops={'facecolor':'gray'})
plt.annotate('', xy=[2, -3], xytext=[2,-1], arrowprops={'facecolor' : 'gray'})

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('함수의 기울기')
plt.show()
```

# 미분가능, 미분 불가능

```python
def ReLU(x) : 
    return np.where(x>0, x, 0)

xx = np.linspace(-5,5,100)
plt.plot(xx, ReLU(xx))
plt.plot(0,0, 'o', ms=10)
plt.title('ReLU 함수 : $x=0$ 에서 미분불가능(불연속)')
plt.xlabel('$x$')
plt.ylabel('$ReLU$')
plt.show()
```

# 미분법칙 

```python
# 원래 함수를 미분법칙에 따라 미분해서 도함수를 구해라. 원래 함수와 그 도함수를 그래프로 나타내자. 

xx = np.linspace(-1,3,1000)
def f(x) : 
    return x**3-3*x**2+x
def fprime(x) : 
    return 3*x**2-6*x+1
fy = f(xx)
fprimey = fprime(xx)

x1, x2 = np.roots([3,-6,1])

plt.subplot(2,1,1)
plt.plot(xx, fy)
plt.title('$f(x)$')
plt.axvline(x1, c='r', ls=':')
plt.axvline(x2, c='r', ls=':')

plt.subplot(2,1,2)
plt.plot(xx, fprimey)
plt.title('1차 도함수 $fprime(x)$')
plt.axvline(x1, c='r', ls=':')
plt.axvline(x2, c='r', ls=':')

plt.tight_layout()
plt.show()
```

```python
xx2 = np.linspace(-20, 40,400)

x1, x2 = np.roots([1, -12, 32])

def f(x) : 
    return (1/3)*x**3-6*x**2+32*x

def ff(x) : 
    return x**2-12*x+32

def fff(x) : 
    return 2*x-12

plt.subplot(3,1,1)
plt.plot(xx2, f(xx2))
plt.title('$f(x)$')
plt.axvline(x2, c='r', ls=':')
plt.axvline(x1, c='r', ls=':')
plt.axhline(0, c='g', ls=':')
plt.ylim(-5000, 10000)

plt.subplot(3,1,2)
plt.plot(xx2, ff(xx2))
plt.axvline(x1, c='r', ls=':')
plt.axvline(x2, c='r', ls=':')
plt.axhline(0, c='g', ls=':')
plt.title("$f'(x)$")
plt.ylim(-10, 1000)

plt.subplot(313)
plt.plot(xx2, fff(xx2))
plt.axhline(0, c='g', ls=':')
plt.title("$f''(x)$")
plt.ylim(-50,50)

plt.tight_layout()
plt.show()
```

```python
# 다시해보자. 
xx = np.linspace(-1, 3, 400)

def f(x) : 
    return x**3-3*x**2+x

def ff(x) : 
    return 3*x**2-6*x+1

def fff(x) : 
    return 6*x-6

x1, x2 = np.roots([3, -6, 1])

plt.figure(figsize=(10,10))

plt.subplot(3,1,1)
plt.plot(xx, f(xx))
plt.xlim(-2,4)
plt.xticks(np.arange(-1,4))
plt.yticks(np.arange(-5,4))
plt.title('함수 f(x)')
plt.xlabel('$x$')
plt.axvline(x1, c='m', ls=':')
plt.axvline(x2, c='m', ls=':')

plt.subplot(3,1,2)
plt.plot(xx, ff(xx))
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-3,11))
plt.title("도함수 f'(x)")
plt.axvline(x1, c='m', ls=':')
plt.axvline(x2, c='m', ls=':')
plt.axhline(0, c='y', ls=':')

plt.subplot(3,1,3)
plt.plot(xx, fff(xx))
plt.xlim(-2,4)
plt.xticks(np.arange(-1,4))
plt.xticks(np.arange(-1,4))
plt.title("2차 도함수 f''(x)")
plt.axhline(0, c='y', ls=':')

plt.tight_layout()
plt.show()
```

# 심파이 미분

$2x^{4}+5x^{3}+6x^{2}+7x+8$ 이 식으로 심파이 미분해보자. 

```python
# 심볼릭 변수 설정 
import sympy
x, y = sympy.symbols('x y')
x
```

```python
f = 2*x**4+5*x**3+6*x**2+7*x+8 # 심볼릭함수 
f
```

```python
# 단변수함수 미분
sympy.diff(f)
```

```python
# 심파이로 심볼릭 함수 미분 
sympy.diff(f, x)
```

```python
# simplify() : 수식정리 함수 
sympy.simplify(sympy.diff(f,x))
```

```python
# 다변수함수 
f = x**2+4*x*y+4*y**2
f
```

```python
# 편미분
sympy.diff(f, x)
```

```python
sympy.diff(f, y)
```

```python
# 심볼릭변수가 여러 개 일 때는 무조건 편미분하듯 어떤 변수로 미분할건지 지정해야 한다. 
x, mu, sigma = sympy.symbols('x, mu, sigma')
f = sympy.exp((x-mu)**2/sigma**2)
f
```

```python
sympy.diff(f, x)
```

```python
sympy.simplify(sympy.diff(f, x))
```

```python
# 2차 도함수 
sympy.diff(f, x, x) # x로 두번 미분
```

```python
sympy.simplify(sympy.diff(f, x,x))
```

# 정적분

# 미적분학의 기본정리로 정적분 계산

```python
x = sympy.symbols('x')

f = x**3-3*x**2+x+6

F = sympy.integrate(f, x)
F # 반-미분해서 찾은 원래 함수 F
```

```python
(F.subs(x,2) - F.subs(x, 0)).evalf()
```

# 수치적분

```python
# 4.3.5 연습문제 
#1. 미적분학의 기본정리로 정적분
x = sympy.symbols('x')
f = 3*x**2-6*x+1

F = sympy.integrate(f, x)
F.subs(x,1)-F.subs(x,0)
```

```python
# 수치적분으로 정적분
def f(x) : 
    return 3*x**2-6*x+1

sp.integrate.quad(f, 0,1) # 수치적분으로 구한 정적분값 

# 미적분학 기본정리 결과와 수치적분 결과가 같게나왔다. 
```

```python
# 2. 
x = sympy.symbols('x')

f2 = 2+6*x+4*sympy.exp(x)+(5/x)
F = sympy.integrate(f2, x) # 원래함수 F

# 미적분학의 기본정리로 정적분
(F.subs(x, 10) - F.subs(x,1)).evalf() # evalf() 메서드로 실젯값 표현 
```

```python
# 수치적분으로 정적분
def f2(x) : 
    return 2+6*x+4*np.exp(x)+(5/x)

sp.integrate.quad(f2, 1, 10)[0] # 수치적분으로 구한 정적분값 

# 미적분학의 기본정리 결과와 수치적분 결과 같았다. 
```

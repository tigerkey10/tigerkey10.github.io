# 좌표변환이란? 
- 벡터공간 기존 기저벡터 선형조합해서 새 기저벡터를 만들 수 있다. 
- 벡터공간 기저벡터 바꿀 때, 좌표도 바꿔주는 작업을 말한다. 
- 변환행렬은 기존 기저벡터에 대한 새 기저벡터 좌표들을 열벡터로 가진 행렬 말한다. 보통 행렬 A라고 부른다. 
- 변환행렬을 사용한다. $A^-1$
- $x_g = A^-1*x_e$ 

```python
#3.2.8 연습문제 
#새 기저벡터 
g1 = np.array([1, 0.75])
g2 = np.array([-1, 0.75])
G = np.vstack([g1,g2]).T

#좌표변환 하려면 A-1 변환행렬이 필요하다
e1 = np.array([1,0])
e2 = np.array([0,1])
E = np.vstack([e1,e2]).T
E #정방행렬이 풀랭크니까 --> 역행렬 존재한다. 

#변환행렬의 역행렬
A = G@np.linalg.inv(E)

#변환행렬
A_1 = np.linalg.inv(A)

a = np.array([1,0])
b = np.array([1,2])
c = np.array([-1,2])

#좌표변환 

print(A_1@a)
print(A_1@b)
print(A_1@c)
```

```python
# 3.2.9 연습문제
g1 = np.array([1, 0.75])
g2 = np.array([-1,0.75])
G = np.vstack([g1,g2]).T

e1 = np.array([0,-1])
e2 = np.array([1,0])
E = np.vstack([e1,e2]).T

A = G@np.linalg.inv(E) #변환행렬의 역행렬

import scipy as sp
import scipy.misc
import scipy.ndimage

f = sp.misc.face(gray=True) #사진 이미지

f1 = sp.ndimage.affine_transform(f, A)

plt.subplot(122)
plt.imshow(f1, cmap=mpl.cm.bone, alpha=0.8)
plt.axis('off')
plt.title('좌표변환 후')
plt.show()
```

```python
# 3.2.9 다시 해보기 
import scipy as sp
import scipy.misc
import scipy.ndimage

xg1 = np.array([1, 0.75])
xg2 = np.array([-1, 0.75])
A = np.vstack([xg1, xg2]).T

e1 = np.array([0,-1])
e2 = np.array([1, 0])
E = np.vstack([e1,e2]).T

g1 = E@A[:,0] #새로운 기저벡터 g1
g2 = E@A[:,1] #새로운 기저벡터 g2

f = sp.misc.face(gray=True)
f2 = sp.ndimage.affine_transform(f, A)
plt.subplot(121)
plt.imshow(f2, cmap=mpl.cm.bone, alpha=0.9)
plt.axis('off')
plt.title('좌표변환 후2')
plt.show()
```

```python
# 연습문제 3.3.1
B = np.array([[2,3],
             [2,1]])

ld = 4
v1 = np.array([3/np.sqrt(13), 2/np.sqrt(13)])

B@v1 == ld*v1

ld2 = -1
v2 = np.array([-1/np.sqrt(2), 1/np.sqrt(2)])

B@v2 == ld2*v2
```

# 고유분해 메서드 np.linalg.eig()

```python
A = np.array([[1, -2], 
             [2, -3]])
ld, V = np.linalg.eig(A)

print(ld)
print(V) #고유벡터행렬, 각 열이 고유벡터다. 
```

```python
B = np.array([[2,3],
             [2,1]])

ld2, V2 = np.linalg.eig(B)
print(ld2)
print(V2)
```

```python
C = np.array([[0, -1],
             [1,0]])

ld3, V3 = np.linalg.eig(C)
print(ld3) #복소수 꼴 고윳값
print(V3)
```

# 심볼릭 연산으로 선형계획법 문제 해결하기 

```python
import cvxpy as cp

a = cp.Variable()
b = cp.Variable()

constraints=[
    a >= 100,
    b >= 100,
    a+2*b <= 500,
    4*a+5*b <= 9800
]

obj = cp.Maximize(3*a+5*b) # 목표
prob = cp.Problem(obj, constraints)

print(prob.solve())
print(prob.status)
print(a.value, b.value)
```

```python
import cvxpy as cp

a = cp.Variable()
b = cp.Variable()

constraints = [
    -a <= -100,
    -b <= -100,
    a+2*b <= 500,
    4*a+5*b <= 9800
]

objective = cp.Minimize(-3*a+-5*b)
problem = cp.Problem(objective, constraints)
print(problem.solve());print(problem.status);print(a.value, b.value)
```

# X,Y 확률변수는 서로의 확률에 전혀 영향 안 미치기 때문에 
# X,Y 확률변수는 서로 독립이다. 

# pgmpy를 이용한 베이지안 확률모형 만들기 

- 사전확률 $P(D)$ & $P(D^{C})$

```python
from pgmpy.factors.discrete import TabularCPD # 조건부확률 모형 구현하는 클래스 
cpd_X = TabularCPD('X',2, np.array([[1-0.002],
                                   [0.002]])) # 한 개 열 원소 합은 1이어야 한다.
print(cpd_X)
```

- 가능도

```python
cpd_Y_on_X = TabularCPD('Y', 2, np.array([[0.95,0.01],
                                         [0.05,0.99]]),
                       evidence=['X'],evidence_card=[2])
print(cpd_Y_on_X)
```

# 이제 베이지안 모형에 사전확률 & 가능도를 넣자. 

```python
from pgmpy.models import BayesianModel

model = BayesianModel([('X','Y')]) # 확률변수 이름 문자열 리스트 # 튜플 형식으로 들어가야 된다. 
model.add_cpds(cpd_X,cpd_Y_on_X)
model.check_model()
```

오케이, 이제 모델이 잘 만들어졌음을 확인했다. 이제 베이지안 정리를 이용해서 사후확률 계산하자. 

```python
from pgmpy.inference import VariableElimination
inference = VariableElimination(model) # 추론 객체 만들고 
post = inference.query(['X'], evidence={'Y':1}) # query() 메서드로 사후확률 구할 확률변수 & 조건으로 쓸 확률변수 명시해줬다.
print(post)
```

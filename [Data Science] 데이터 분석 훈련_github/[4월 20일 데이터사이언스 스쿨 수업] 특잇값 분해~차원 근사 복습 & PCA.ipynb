```python
#PCA
from sklearn.datasets import load_iris
iris = load_iris()
N = 10 #앞의 10송이 데이터(벡터10개만) 선택
X = iris.data[:N,:2] #꽃받침 길이와 꽃받침 폭만 선택
```

```python
iris.data[:N,:2] #이런 꼴이다. 벡터 1개가 데이터 레코드 1개다.
```

```python
plt.plot(X.T,'o:') #각 레코드가 열벡터로 표시된 행렬을 점선 표시
plt.xticks(range(2), ['꽃받침 길이','꽃받침 폭'])
plt.xlim(-0.5, 2) #x축 범위 제한
plt.ylim(2.5,6) #y축 범위 제한
plt.title('붓꽃 크기 특성')
plt.legend(['표본 {}'.format(i+1) for i in range(N)])
plt.show()
```
